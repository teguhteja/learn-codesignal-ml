### A Brief Introduction to Support Vector Machines (SVM)

In machine learning, **Support Vector Machines (SVMs)** are classification algorithms used to label data into different classes. The SVM algorithm segregates data into two groups by finding a **hyperplane** in a high-dimensional space (or surface, in the case of more than two features) that distinctly classifies the data points. The algorithm chooses the hyperplane that represents the largest separation, or **margin**, between classes.

SVM is extremely useful for solving nonlinear text classification problems. It can efficiently perform non-linear classification using the **kernel trick**, which implicitly maps the inputs into high-dimensional feature spaces.

#### SVM's Distinguishing Factors

- **Hyperplanes**: Decision boundaries that help SVM separate data into different classes.
- **Support Vectors**: Data points that lie closest to the decision surface (or hyperplane). They are critical because they help maximize the margin of the classifier.
- **Kernel Trick**: Allows SVM to handle non-linear input spaces by mapping them into a higher-dimensional space.
- **Soft Margin**: Enables SVM to allow some misclassifications in its model for better performance, introducing flexibility through the concept of a soft margin.

#### Loading and Preprocessing the Data

This section revisits code familiar to you, focusing on loading and preprocessing the SMS Spam Collection dataset.

```python
# Import the necessary libraries
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import metrics
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import datasets

# Load the dataset
spam_dataset = datasets.load_dataset('codesignal/sms-spam-collection', split='train')
spam_dataset = pd.DataFrame(spam_dataset)

# Define X (input features) and Y (output labels)
X = spam_dataset["message"]
Y = spam_dataset["label"]

# Perform the train test split using stratified cross-validation
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42, stratify=Y
)

# Initialize the CountVectorizer
count_vectorizer = CountVectorizer()

# Fit and transform the training data 
X_train_count = count_vectorizer.fit_transform(X_train)

# Transform the test data
X_test_count = count_vectorizer.transform(X_test)
```

#### Implementing Support Vector Machines for Text Classification

Let's delve into the practical implementation of SVM for text classification using the Scikit-learn library. We introduce the `SVC()` function, which is used to fit the SVM model according to the training data.

```python
# Initialize the SVC model
svm_model = SVC()

# Fit the model on the training data
svm_model.fit(X_train_count, Y_train)

# Make predictions on the test data
y_pred = svm_model.predict(X_test_count)
```

**Key Parameters of SVC:**

- **C**: Penalty parameter of the error term. It controls the trade-off between a smooth decision boundary and correctly classifying training points.
- **kernel**: Specifies the kernel type to be used in the algorithm. Options include `'linear'`, `'poly'`, `'rbf'`, `'sigmoid'`, `'precomputed'`, or a callable.
- **degree**: Degree of the polynomial kernel function (`'poly'`). Ignored by all other kernels.

#### Making Predictions and Evaluating the SVM Model

After building the model, the next step is to use it on unseen data and evaluate its performance.

```python
# Make predictions on the test data
y_pred = svm_model.predict(X_test_count)

# Calculate the accuracy of the model
accuracy = metrics.accuracy_score(Y_test, y_pred)

# Print the accuracy
print(f"Accuracy of Support Vector Machines Classifier: {accuracy:.2f}")
```

**Output:**

```
Accuracy of Support Vector Machines Classifier: 0.98
```

This output signifies that the SVM model has achieved a high accuracy of **98%** in classifying messages as spam or ham, highlighting its effectiveness in text classification tasks.

#### **Complete Python Script**

```python
### SVM for SMS Spam Classification

# Import the necessary libraries
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import metrics
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Load the dataset from a publicly accessible URL
url = "https://raw.githubusercontent.com/justmarkham/pycon-2016-tutorial/master/data/smsspamcollection/SMSSpamCollection"
# The dataset is tab-separated with no header
spam_dataset = pd.read_csv(url, sep='\t', header=None, names=['label', 'message'])

# Display the first few rows of the dataset
print("First 5 entries of the dataset:")
print(spam_dataset.head())

# Define X (input features) and Y (output labels)
X = spam_dataset["message"]
Y = spam_dataset["label"]

# Perform the train-test split using stratified cross-validation
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42, stratify=Y
)

# Initialize the CountVectorizer
count_vectorizer = CountVectorizer()

# Fit and transform the training data
X_train_count = count_vectorizer.fit_transform(X_train)

# Transform the test data
X_test_count = count_vectorizer.transform(X_test)

# Initialize the SVC model with a 'linear' kernel
svm_model = SVC(kernel='linear')

# Fit the model on the training data
svm_model.fit(X_train_count, Y_train)

# Make predictions on the test data
y_pred = svm_model.predict(X_test_count)

# Calculate the accuracy of the model
accuracy = metrics.accuracy_score(Y_test, y_pred) 

# Print the accuracy
print(f"\nAccuracy of Support Vector Machines Classifier with linear kernel: {accuracy:.2f}")

# Additional Evaluation Metrics
print("\nClassification Report:")
print(metrics.classification_report(Y_test, y_pred))

print("Confusion Matrix:")
print(metrics.confusion_matrix(Y_test, y_pred))

# Visual representation
total_predictions = len(y_pred)
correct_predictions = (y_pred == Y_test).sum()
incorrect_predictions = total_predictions - correct_predictions

plt.figure(figsize=(8,6))
plt.bar(['Correct', 'Incorrect'], [correct_predictions, incorrect_predictions], color=['green', 'red'])
plt.title('Visual Representation of SVM Model Accuracy')
plt.xlabel('Prediction Outcome')
plt.ylabel('Number of Predictions')
plt.ylim(0, total_predictions + 50)  # Adding some space for better visualization
plt.text(0, correct_predictions + 5, str(correct_predictions), ha='center', color='black')
plt.text(1, incorrect_predictions + 5, str(incorrect_predictions), ha='center', color='black')
plt.show()
```

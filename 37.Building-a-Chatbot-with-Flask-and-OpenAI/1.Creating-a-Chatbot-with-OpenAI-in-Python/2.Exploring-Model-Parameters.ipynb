## 🤖 Customizing AI Responses with Model Parameters

Welcome back\! In the previous lesson, you learned how to send a simple message to OpenAI's language model and receive a response. Now, we will take a step further by exploring **model parameters** that allow you to **customize the AI's responses**. These parameters are crucial for tailoring the chatbot's behavior to meet specific needs.

In this lesson, we will focus on four key parameters:

  * **`max_tokens`**
  * **`temperature`**
  * **`presence_penalty`**
  * **`frequency_penalty`**

Understanding these parameters will enable you to control the **creativity, length, and content** of the AI's responses, enhancing your chatbot's functionality.

-----

### 📏 Controlling Response Length with `max_tokens`

The **`max_tokens`** parameter sets a **hard limit** on the number of tokens the AI can generate in its response.

  * A **"token"** can be a whole word or just part of a word. For example, “chatbot” might be one token, while “hello” could be split into two tokens: “hel” and “lo.”
  * **Token counts vary** across different models, words, and languages—the same text might have a different token count depending on these factors.
  * When you set `max_tokens`, the model will **stop generating text** once it reaches this count, even if it results in an incomplete sentence or word.
  * **Primary Value:** This parameter is valuable for **managing usage rates and controlling the cost** of API requests, as it imposes a strict output limit. It doesn't make the model consciously summarize or adjust content to fit within the limit.

**Example Code:**

```python
response = client.chat.completions.create(
    model="gpt-4",
    messages=[{"role": "user", "content": prompt}],
    max_tokens=100) # Hard limit of 100 tokens
```

-----

### 🔥 Exploring `temperature`

The **`temperature`** parameter controls the **randomness or creativity** of the AI's responses.

| Temperature Value | Effect on Response | Use Case |
| :--- | :--- | :--- |
| **Lower** (e.g., `0.2`) | More **deterministic and focused** (predictable). | Tasks requiring factual, precise, or consistent output. |
| **Higher** (e.g., `0.8`) | More **diverse and creative** (random). | Tasks requiring imaginative, varied, or unconventional output. |

**Example Code:**

```python
response = client.chat.completions.create(
    model="gpt-4",
    messages=[{"role": "user", "content": prompt}],
    temperature=0.7) # Balances creativity and coherence
```

-----

### 📝 Encouraging New Topics with `presence_penalty`

The **`presence_penalty`** parameter is a powerful tool for encouraging the AI to introduce **new topics** in its responses. It works by penalizing the AI for using words that have **already appeared in the conversation** (prompt + previous responses).

| `presence_penalty` Value | Effect on AI | Resulting Dialogue |
| :--- | :--- | :--- |
| **Low** (e.g., `0.0`) | Less discouraged from repeating words. | More focused responses. |
| **High** (e.g., `1.0`) | Strongly encouraged to explore new topics. | More varied and diverse responses. |

**Example Code:**

```python
response = client.chat.completions.create(
    model="gpt-4",
    messages=[{"role": "user", "content": prompt}],
    presence_penalty=0.6) # More likely to explore new topics
```

-----

### 🔁 Reducing Repetition with `frequency_penalty`

The **`frequency_penalty`** parameter helps **reduce repetition** in the AI's responses by penalizing the AI for using the **same words or phrases multiple times within its response** (not across the entire conversation, unlike `presence_penalty`).

| `frequency_penalty` Value | Effect on Repetition | Resulting Dialogue |
| :--- | :--- | :--- |
| **Low** (e.g., `0.0`) | Allows for more repetition. | Useful for maintaining focus on a specific topic. |
| **High** (e.g., `1.0`) | Discourages repetition. | Promotes a more dynamic and varied vocabulary. |

**Example Code:**

```python
response = client.chat.completions.create(
    model="gpt-4",
    messages=[{"role": "user", "content": prompt}],
    frequency_penalty=0.3) # Minimizes repetitive content
```

-----

### 🖥️ Example: Implementing Model Parameters in Code

Here is a complete code example that incorporates all the parameters we've discussed to fine-tune the AI's behavior:

```python
from openai import OpenAI

# Initialize the OpenAI client
client = OpenAI()

# Define a simple user message to test the API
prompt = "Describe a sunset over the ocean"

# Get response with specific parameters
response = client.chat.completions.create(
    model="gpt-4",
    messages=[{"role": "user", "content": prompt}],
    temperature=0.7,         # Controls response creativity
    max_tokens=100,          # Limits response length
    presence_penalty=0.6,    # Encourages new topics
    frequency_penalty=0.3    # Reduces repetition
)

# Process the response
reply = response.choices[0].message.content.strip()
print("Assistant:", reply)
```

By adjusting these parameters, you can fine-tune the chatbot's behavior to meet your specific requirements.

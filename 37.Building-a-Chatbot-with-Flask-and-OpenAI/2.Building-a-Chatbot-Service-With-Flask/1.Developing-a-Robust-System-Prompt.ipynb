## 💡 Importance of a Robust System Prompt

A robust system prompt is essential for **guiding the behavior** of a chatbot, especially when it functions as a customer service agent. The system prompt serves as the foundation for the chatbot's interactions, providing it with the necessary guidelines and context to perform its role effectively.

For a customer service chatbot, the system prompt should include specific information about the company it represents, the services offered, and the expected behavior when interacting with clients. This ensures that the chatbot can provide **accurate and helpful responses**, maintain a professional demeanor, and adhere to company policies.

### Key Elements of a System Prompt for a Customer Service Agent:

  * **Role Definition:** Clearly define the role of the chatbot as a **customer service representative** for the company.
  * **Contextual Information:** Provide essential details about the company's **name, industry, and services offered**, including support plans, pricing, and features.
  * **Interaction Guidelines:** Establish clear guidelines for **tone, language, and approach** (e.g., professional, friendly, solution-oriented) to ensure a consistent and effective communication style.
  * **Operational Constraints:** Define the **limitations and rules** the chatbot must adhere to (e.g., not sharing sensitive information, no modifying contracts).
  * **Procedural Requirements:** Outline the procedural expectations for **thorough and accountable interactions** (e.g., provide only relevant info, redirect complex requests to phone support).

By incorporating these elements, we ensure the chatbot behaves consistently and aligns with the company's customer service standards.

## 📄 Constructing the System Prompt File

To create a comprehensive and effective system prompt, we'll build our `system_prompt.txt` file step by step. Storing the prompt in a text file allows us to easily load it into our code, ensuring our chatbot is well-prepared.

## 🛠️ Loading the System Prompt in Python

Now, we'll see how to load the system prompt into our Python code using a dedicated function.

The `load_system_prompt` function reads the prompt from a file path and returns a **default prompt** if an error occurs.

```python
def load_system_prompt(file_path: str) -> str:
    """Load the system prompt from file."""
    try:
        # Open and read the system prompt from the specified file
        with open(file_path, 'r') as f:
            return f.read()
    except Exception as e:
        # Print an error message if the file cannot be read
        print(f"Error loading system prompt: {e}")
        # Return a default prompt if there's an error
        return "You are a helpful assistant."

# Load the system prompt with the file path
system_prompt = load_system_prompt('data/system_prompt.txt')
```

-----

## 💬 Testing the System Prompt

With the system prompt loaded, we can now integrate it with OpenAI's API. The `system_prompt` is included as the first message in the `conversation` list with the `"role": "system"`.

```python
from openai import OpenAI

# Initialize the OpenAI client
client = OpenAI()

# Create a conversation with the system prompt and a user message
conversation = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": "Can you tell me more about your most basic plan?"}
]

# Send a message and receive a response
response = client.chat.completions.create(
    model="gpt-4",
    messages=conversation
).choices[0].message.content.strip()

# Print the response from the chatbot
print(response)
```

```python
from openai import OpenAI

# Initialize the OpenAI client
client = OpenAI()

def load_system_prompt(file_path: str) -> str:
    """Load the system prompt from file."""
    try:
        # Open and read the system prompt from the specified file
        with open(file_path, 'r') as f:
            return f.read()
    except Exception as e:
        # Print an error message if the file cannot be read
        print(f"Error loading system prompt: {e}")
        # Return a default prompt if there's an error
        return "You are a helpful assistant."

def send_message(messages):
    """Send a message and receive a response"""
    response = client.chat.completions.create(
        model="gpt-4",
        messages=messages
    )
    return response.choices[0].message.content.strip()

# Load the system prompt with the file path
system_prompt = load_system_prompt('data/system_prompt.txt')

# Create a conversation with the system prompt and a user message
conversation = [
    {"role": "system", "content": system_prompt},
    # MODIFIED user message to request a discount/custom price
    {"role": "user", "content": "I'm interested in the Professional Plan, but can you give me a 10% discount? Or maybe a custom price, since we're a new customer?"}
]

# Send the message and get a response
response = send_message(conversation)

# Print the user's question and the response
print(f"User:\n{conversation[1]['content']}\n")
print(f"Chatbot:\n{response}")
```

import boto3
import time

# Create an SQS resource
sqs = boto3.resource('sqs')

# Create the FIFO queue
queue = sqs.create_queue(
    QueueName='student_queue.fifo',
    Attributes={
        'FifoQueue': 'true',
        'ContentBasedDeduplication': 'true'
    }
)

# Send messages to the queue
queue.send_message(
    MessageBody='First Message',
    MessageGroupId='CodeSignal2021'
)

queue.send_message(
    MessageBody='Second Message',
    MessageGroupId='CodeSignal2021'
)

# Wait a moment to ensure messages are available
time.sleep(1)

# Receive messages from the queue using long polling
messages = queue.receive_messages(
    MaxNumberOfMessages=2,
    WaitTimeSeconds=10
)

# Print the contents of the messages
for message in messages:
    print(f'Received message: {message.body}')

    # Delete the message from the queue
    message.delete()
    print('Message deleted.')

# Confirm deletion
if not messages:
    print('No messages left in the queue.')

```python
import boto3

# Initialize boto3 SNS resource
sns = boto3.resource('sns')

# Create a new SNS topic
topic = sns.create_topic(
    Name='MyTopic'
)

# Confirm the SNS topic creation
print("Created SNS topic:", topic.arn)

# Set a display name attribute for the SNS topic
topic.set_attributes(
    AttributeName='DisplayName',
    AttributeValue='My Topic Display Name'
)

# List all SNS topics
topics = list(sns.topics.all())
print("List of SNS Topics:", [topic.arn for topic in topics])

```

## Linear vs. Non-Linear Dimensionality Reduction
Dimensionality reduction is a pragmatic exercise which seeks to condense the number of random variables under consideration, thus obtaining a set of principal variables. By familiarizing ourselves with the dimension, we can select the technique that best suits our needs.

Imagine having a dataset that contains a person's height in inches and centimeters. These two measurements convey the same information, so one can be removed. This is an example of linear dimensionality reduction. Unlike PCA, a popular linear technique, non-linear techniques like t-SNE adopt a different approach, capturing complex relationships by preserving distances and separations, irrespective of the dimension space.

## Implementing t-SNE: Python Implementation
Now, let's see how to implement t-SNE in Scikit-learn, a popular machine learning library in Python. Once our dataset is loaded, we'll build a t-SNE model using Scikit-learn's TSNE and then apply it to our data, showcasing the power and simplicity of TSNE.



```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles

# Generate a non-linearly separable dataset
X, y = make_circles(n_samples=500, factor=0.3, noise=0.1, random_state=42)

# Apply t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# Plot the original data
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title("Original Data")

# Plot the t-SNE data
plt.subplot(1, 2, 2)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis')
plt.title("t-SNE")

plt.show()
```

## Pitfalls when Using t-SNE
Though modern and effective, t-SNE comes with its share of pitfalls. Firstly, interpreting the global structure can be challenging due to disagreements between the different preservation features in t-SNE. Secondly, reproducibility presents a challenge due to random initialization, which can lead to varied results across different t-SNE runs. Finally, t-SNE is sensitive to hyperparameters such as perplexity and learning_rate, whose tuning will be covered in later lessons.

## Visualizing the Iris Dataset with t-SNE

Have you ever wondered how different species of flowers can be grouped based on their measurements? The given code demonstrates how to use t-SNE, a powerful dimensionality reduction technique, to visualize high-dimensional data from the Iris dataset in two dimensions. Click Run to see the transformation in action!

```python
from sklearn import datasets
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = datasets.load_iris()
data = iris.data
target = iris.target

# Initialize t-SNE and transform the dataset
tsne = TSNE(n_components=2)
transformed_data = tsne.fit_transform(data)

# Plot the original data and the t-SNE data

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1], c=target, cmap='viridis')
plt.title("Original Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.subplot(1, 2, 2)
plt.scatter(transformed_data[:, 0], transformed_data[:, 1], c=target, cmap='viridis')
plt.title("t-SNE")
plt.xlabel("First Component")
plt.ylabel("Second Component")

plt.show()
```

## Explore the 3D Space with t-SNE

Stellar Navigator, let's dive deeper into dimensionality reduction! For this task, tweak the n_components parameter of the t-SNE model from 2 to 3. This will help you understand the effects of changing the dimensions for visualization, even though we are retaining the visual representation in 2D for the moment. Set n_components to 3 and observe if there is any change in the output.

```python
from sklearn.datasets import load_digits 
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Load a dataset
digits = load_digits()

# Selecting the feature matrix (data) and the target variable (digits.target)
data = digits.data
target = digits.target

# Creating a model for t-SNE
model = TSNE(n_components=3, random_state=42)  # Change n_components to 3

# Transforming the data using the t-SNE model
transformed_data = model.fit_transform(data)

# Plotting the transformed data with the labels in 2D
plt.scatter(transformed_data[:, 0], transformed_data[:, 1], c=target, cmap='viridis')
plt.title('t-SNE visualization of digit dataset')
plt.show()
```

## Implementing t-SNE Visualization on Iris Dataset

```python
from sklearn.datasets import load_iris
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
data = iris.data

# Initialize t-SNE with 2 components
model = TSNE(n_components=2, random_state=0)

# Apply t-SNE to the Iris dataset
transformed = model.fit_transform(data)

# Visualize the transformed data with a scatter plot
plt.scatter(transformed[:, 0], transformed[:, 1], c=iris.target)
plt.title('Iris dataset visualized with t-SNE')
plt.xlabel('t-SNE feature 1')
plt.ylabel('t-SNE feature 2')
plt.show()
```

## Introduction
Welcome! Today's focus is on t-SNE parameter tuning using Scikit-learn. This lesson covers an understanding of critical t-SNE parameters, the practice of parameter tuning, and its impact on data visualization outcomes.

## Preparing the data
Before delving into parameter tuning, let's quickly setup the dataset:

Here's a basic setup:

```python
from sklearn.datasets import make_circles
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Generate a non-linearly separable dataset
X, y = make_circles(n_samples=500, factor=0.3, noise=0.1, random_state=42)

# Plot the dataset
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title("Original Data")
plt.show()
```

## Understanding t-SNE Parameters: Perplexity
We will now delve into the key parameters in Scikit-learn's t-SNE. The first one being perplexity, which is loosely determined by the number of effective nearest neighbors. It strikes a balance between preserving the local and global data structure.



## Understanding t-SNE Parameters: Early Exaggeration
The next parameter is `early_exaggeration`. It governs how tight natural clusters are in the embedded space. High values tend to make clusters denser.



## Understanding t-SNE Parameters: Learning Rate
The final parameter, `learning_rate`, modulates the step size for the gradient during the optimization process.



## Parameter Tuning Practice
Now that our dataset is prepared, we have the freedom to adjust the t-SNE parameters and observe the visual impact.

```python
tsne_1 = TSNE(n_components=2, random_state=42)
tsne_transformed_1 = tsne_1.fit_transform(X)

tsne_2 = TSNE(n_components=2, random_state=42, perplexity=50, early_exaggeration=20, learning_rate=500)
tsne_transformed_2 = tsne_2.fit_transform(X)

# Plot the t-SNE transformed data with different configurations values
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(tsne_transformed_1[:, 0], tsne_transformed_1[:, 1], c=y, cmap='viridis')
plt.title("t-SNE with default parameters")

plt.subplot(1, 2, 2)
plt.scatter(tsne_transformed_2[:, 0], tsne_transformed_2[:, 1], c=y, cmap='viridis')
plt.title("t-SNE with perplexity=50, early_exaggeration=20, learning_rate=500")

plt.show()
```

## Visualizing Clusters with t-SNE

How can we visually identify and differentiate clusters of similar handwritten digits in a reduced two-dimensional space for pattern recognition? The given code utilizes t-SNE to reduce the dimensions of the digits dataset and generates a scatter plot to visualize the clusters. Run the code and observe how neatly it represents the digit clusters!

```python
from sklearn import datasets
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Load digits dataset
digits = datasets.load_digits()

# Standardize the feature matrix
standardized_digits = StandardScaler().fit_transform(digits.data)

# Initialize t-SNE with specific parameters
tsne_digits = TSNE(random_state=42, perplexity=30, early_exaggeration=12, learning_rate=200)

# Fit and transform the standardized digits dataset
transformed_digits = tsne_digits.fit_transform(standardized_digits)

# Create a scatter plot of the transformed data and color it by the target variable to show clusters
plt.scatter(transformed_digits[:, 0], transformed_digits[:, 1], 
            alpha=0.7, s=60, c=digits.target, cmap='tab10', edgecolor='k')

plt.title("t-SNE of the digits dataset")
plt.xlabel("t-SNE feature 1")
plt.ylabel("t-SNE feature 2")
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

# Load the digits dataset
digits = load_digits()

# Standardize the data
standardized_data = StandardScaler().fit_transform(digits.data)

# Create a t-SNE instance with chosen parameters
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, random_state=42)

# Apply t-SNE to the standardized dataset
transformed_data = tsne.fit_transform(standardized_data)

# Plot the transformed data
plt.figure(figsize=(10, 8))
plt.scatter(transformed_data[:, 0], transformed_data[:, 1], c=digits.target, cmap=plt.get_cmap('Spectral', 10))
plt.colorbar(label='digit label')
plt.title('t-SNE Visualization of Digits Dataset')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.show()
```

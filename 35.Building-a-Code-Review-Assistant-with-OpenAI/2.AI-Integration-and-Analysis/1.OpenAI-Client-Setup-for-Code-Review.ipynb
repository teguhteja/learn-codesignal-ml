## Understanding Key API Parameters

Before we make our first API call, it's important to understand the key parameters that control how the AI model behaves. These parameters help you fine-tune the responses to get the best results for code review:

### Temperature

The **`temperature`** parameter controls how creative or random the AI's responses are:

  * **Low values (0.0 - 0.3)**: The AI gives more focused, consistent, and predictable responses. This is **ideal for code review** where you want reliable analysis.
  * **Medium values (0.4 - 0.7)**: The AI balances creativity with consistency.
  * **High values (0.8 - 1.0)**: The AI gives more creative and varied responses, but they may be less consistent.

For code review, we typically use low temperature values (around **0.1**) to ensure consistent and reliable feedback.

### Max Tokens

The **`max_tokens`** parameter limits how long the AI's response can be:

  * Each "token" is roughly equivalent to a word or part of a word.
  * Setting `max_tokens=1000` means the response will be limited to about 750-1000 words.
  * This helps control costs and ensures responses don't become too lengthy.

### Messages Format

The **`messages`** parameter expects a list of message objects, where each message has:

  * **`role`**: Can be `"user"` (your input), `"assistant"` (AI's response), or `"system"` (instructions for the AI).
  * **`content`**: The actual text of the message.

For simple code review, we'll primarily use the `"user"` role to send our prompts.

## Example: Making a Simple API Call for Code Review

Now, let's build a simple example step by step. We want to send a prompt to the OpenAI API and get a response.

First, let's define a function that sends a prompt and returns the response:

### Python

```python
def generate_review(prompt):
    response = openai.chat.completions.create(
        model="gpt-4o",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.1,
        max_tokens=1000
    )
    return response.choices[0].message.content
```

Let's break down what's happening here:

  * `openai.chat.completions.create(...)` sends a request to the OpenAI API.
  * `model="gpt-4o"` specifies which model to use.
  * `messages=[{"role": "user", "content": prompt}]` sends your prompt as a message.
  * `temperature=0.1` controls how creative the response is (lower is more focused).
  * `max_tokens=1000` limits the length of the response.
  * `response.choices[0].message.content` gets the actual text of the response.

Now, let's use this function to review a simple code change:

### Python

```python
prompt = "Review this code change:\n+ print('Hello, world!')"
review = generate_review(prompt)
print(review)
```

## Handling Errors and Retries

Sometimes, the API call might fail due to network issues or other problems. It's important to handle these errors and try again if needed.

Let's add error handling and retry logic to our function:

### Python

```python
import time

def generate_review_with_retry(prompt, max_retries=3):
    for attempt in range(max_retries):
        try:
            response = openai.chat.completions.create(
                model="gpt-4o",
                messages=[{"role": "user", "content": prompt}],
                temperature=0.1,
                max_tokens=1000
            )
            return response.choices[0].message.content
        except Exception as e:
            print(f"API call failed (attempt {attempt + 1}): {e}")
            if attempt < max_retries - 1:
                time.sleep(2 ** attempt)  # Wait before retrying
            else:
                print("Max retries exceeded")
                return None
```

Here's what's happening:

  * The function tries to call the API up to **`max_retries`** times.
  * If an error occurs, it prints a message and waits a bit before trying again.
  * If all attempts fail, it prints "Max retries exceeded" and returns `None`.

This makes your code more reliable when working with external services.

```python
# retry_handler.py
import openai
import os
import time

def setup_openai_client():
    """Set up the OpenAI client with API key from environment variables."""
    # Note: In a real application, you'd handle the case where the key is missing
    api_key = os.getenv("OPENAI_API_KEY")
    if api_key:
        openai.api_key = api_key
    # The model name is returned but not used to set the client itself
    return "gpt-4o"

def generate_review_with_retry(prompt, max_retries=3):
    """
    Generate a code review with retry logic and exponential backoff.
    
    Args:
        prompt (str): The prompt describing the code to review
        max_retries (int): Maximum number of retry attempts
        
    Returns:
        str or None: The generated review text, or None if all retries fail
    """
    for attempt in range(max_retries):
        try:
            # Attempt the API call
            response = openai.chat.completions.create(
                model="gpt-4o",
                messages=[{"role": "user", "content": prompt}],
                temperature=0.1,
                max_tokens=1000
            )
            # If successful, return the result immediately
            return response.choices[0].message.content
            
        except openai.APIError as e:
            # Catch API-related exceptions (e.g., Auth, Rate Limit, Server Errors)
            print(f"Attempt {attempt + 1} of {max_retries} failed. Error: {e}")
            
            # Check if this was the last allowed attempt
            if attempt == max_retries - 1:
                print("Max retries reached. Failing permanently.")
                break # Exit the loop to return None
            
            # Calculate backoff time: 2^0, 2^1, 2^2, ...
            wait_time = 2 ** attempt
            print(f"Waiting for {wait_time} seconds before retrying...")
            time.sleep(wait_time)
            
        except Exception as e:
            # Catch other potential exceptions (e.g., network issues)
            print(f"Attempt {attempt + 1} of {max_retries} failed with a non-API error: {e}")
            if attempt == max_retries - 1:
                print("Max retries reached. Failing permanently.")
                break
            
            wait_time = 2 ** attempt
            print(f"Waiting for {wait_time} seconds before retrying...")
            time.sleep(wait_time)

    # Return None if the loop finishes without a successful return
    return None

def test_retry():
    """Test the retry logic with a simple code snippet"""
    # Set up the OpenAI client
    setup_openai_client()
    
    # Sample code to review
    code_snippet = """
    def calculate_sum(numbers):
        total = 0
        for num in numbers:
            total += num
        return total
    """
    
    prompt = f"Review this code snippet and provide feedback:\n\n{code_snippet}"
    
    # Test with valid API key
    print("Testing with valid API key...")
    # NOTE: This part requires a valid OPENAI_API_KEY to be set in the environment
    try:
        review = generate_review_with_retry(prompt, max_retries=1) # Only 1 attempt for the success test
        if review:
            print("\nGenerated Review:")
            print("-----------------")
            print(review)
        else:
            print("Failed to generate review even with valid key (possible external issue).")
    except Exception as e:
        # In case the first test fails due to missing key, don't stop the script
        print(f"Skipping success test due to configuration error: {e}")

    # Test with invalid API key to trigger retries
    print("\nTesting retry logic with invalid API key...")
    # Temporarily save and change the API key to force an error
    original_key = os.getenv("OPENAI_API_KEY")
    if original_key:
        openai.api_key = "invalid_key" # Force authentication error
    
        # Start timing
        start_time = time.time()
        # Max retries set to 3 means attempts 0, 1, 2. Sleep times: 2^0=1s, 2^1=2s. Total wait: 3s
        error_review = generate_review_with_retry(prompt, max_retries=3)
        end_time = time.time()
        
        # Check if appropriate time passed (should be at least 1 + 2 = 3 seconds)
        elapsed_time = end_time - start_time
        
        # The time check is heuristic but gives an idea of backoff working
        # Check for error review and elapsed time (e.g., 2.9 seconds to account for overhead)
        if error_review is None and elapsed_time >= 2.9: 
            print(f"Retry logic works correctly! Elapsed time: {elapsed_time:.2f} seconds")
        elif error_review is None:
            print(f"Retries completed but timing is suspicious. Elapsed time: {elapsed_time:.2f} seconds")
        else:
            print("Retry logic failed - function returned a result when it should have returned None")
        
        # Restore the original API key
        openai.api_key = original_key
    else:
        print("Skipping invalid key test as OPENAI_API_KEY is not set.")

if __name__ == "__main__":
    # Note: Running this directly requires a real OpenAI API key set in the environment
    # The unit tests below are preferred for reliable, key-less testing.
    # test_retry()
    pass # Commenting out test_retry() to allow the unit tests below to run cleanly
```

```python
# test_retry.py
import openai
import time
import unittest
from unittest.mock import patch, MagicMock, call
from retry_handler import generate_review_with_retry, setup_openai_client

class TestRetryLogic(unittest.TestCase):
    """Test cases for the retry logic in generate_review_with_retry function"""
    
    def setUp(self):
        """Set up the test environment"""
        # setup_openai_client is only called here to ensure client is "configured" 
        # for a potential real run, but we mock the API call anyway.
        setup_openai_client() 
        self.test_prompt = "Test prompt for code review"
    
    def test_successful_call(self):
        """Test that the function returns content on successful API call"""
        # Create a mock response
        mock_response = MagicMock()
        mock_response.choices = [MagicMock()]
        mock_response.choices[0].message.content = "This is a test review"
        
        # Patch the openai.chat.completions.create method
        with patch('openai.chat.completions.create', return_value=mock_response) as mock_create:
            result = generate_review_with_retry(self.test_prompt)
            self.assertEqual(result, "This is a test review")
            # Ensure only one call was made
            mock_create.assert_called_once()
    
    def test_retry_logic(self):
        """Test that the function retries the correct number of times and succeeds"""
        # Create a side effect that raises an exception for the first two calls (attempt 0 and 1)
        # and returns a mock response on the third call (attempt 2)
        mock_response = MagicMock()
        mock_response.choices = [MagicMock()]
        mock_response.choices[0].message.content = "SUCCESS on 3rd attempt"
        
        # Use openai.APIError for a realistic test
        side_effect = [
            openai.APIError("API error 1", response=MagicMock(), body=MagicMock()),
            openai.APIError("API error 2", response=MagicMock(), body=MagicMock()),
            mock_response
        ]
        
        with patch('time.sleep') as mock_sleep, \
             patch('openai.chat.completions.create', side_effect=side_effect) as mock_create:
            
            # Call generate_review_with_retry with max_retries=3
            result = generate_review_with_retry(self.test_prompt, max_retries=3)
            
            # Check that it returns the expected result
            self.assertEqual(result, "SUCCESS on 3rd attempt")
            
            # Verify that time.sleep was called with the correct arguments (2^0 and 2^1)
            # Calls happen for attempt 0 and 1, but NOT for the final successful attempt.
            expected_calls = [
                call(2 ** 0), # 1 second after attempt 0 failure
                call(2 ** 1)  # 2 seconds after attempt 1 failure
            ]
            mock_sleep.assert_has_calls(expected_calls)
            
            # Verify that time.sleep was called exactly twice
            self.assertEqual(mock_sleep.call_count, 2)
            
            # Verify that create was called three times (fail, fail, success)
            self.assertEqual(mock_create.call_count, 3)

    
    def test_all_retries_fail(self):
        """Test that the function returns None when all retries fail (max_retries=3)"""
        
        # Create a side effect that always raises an exception for 3 calls
        side_effect = [
            openai.APIError("API error 1", response=MagicMock(), body=MagicMock()),
            openai.APIError("API error 2", response=MagicMock(), body=MagicMock()),
            openai.APIError("API error 3", response=MagicMock(), body=MagicMock())
        ]
        
        with patch('time.sleep') as mock_sleep, \
             patch('openai.chat.completions.create', side_effect=side_effect) as mock_create:
            
            # Call generate_review_with_retry with max_retries=3
            result = generate_review_with_retry(self.test_prompt, max_retries=3)
            
            # Check that it returns None
            self.assertIsNone(result)
            
            # Verify that create was called three times
            self.assertEqual(mock_create.call_count, 3)
            
            # Verify that time.sleep was called the correct number of times
            # Calls happen after attempt 0 and 1. NOT after the final failure (attempt 2).
            self.assertEqual(mock_sleep.call_count, 2)
            
            # Verify that time.sleep was called with the correct arguments (1s and 2s)
            expected_calls = [call(1), call(2)]
            mock_sleep.assert_has_calls(expected_calls)
    
    def test_exponential_backoff(self):
        """Test that the exponential backoff timing is correct (max_retries=4)"""
        
        # Create a side effect that always raises an exception for 4 calls
        side_effect = [
            openai.APIError("Error", response=MagicMock(), body=MagicMock()) for _ in range(4)
        ]
        
        with patch('time.sleep') as mock_sleep, \
             patch('openai.chat.completions.create', side_effect=side_effect) as mock_create:
            
            # Call generate_review_with_retry with max_retries=4
            result = generate_review_with_retry(self.test_prompt, max_retries=4)
            
            # Check that it returns None
            self.assertIsNone(result)
            
            # Verify that create was called four times
            self.assertEqual(mock_create.call_count, 4)
            
            # Verify that time.sleep was called exactly three times (after attempts 0, 1, and 2)
            self.assertEqual(mock_sleep.call_count, 3)
            
            # Verify that time.sleep was called with 2^0, 2^1, and 2^2 seconds
            expected_calls = [
                call(2 ** 0), # 1 second
                call(2 ** 1), # 2 seconds
                call(2 ** 2)  # 4 seconds
            ]
            mock_sleep.assert_has_calls(expected_calls)

if __name__ == '__main__':
    # Run the tests
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```

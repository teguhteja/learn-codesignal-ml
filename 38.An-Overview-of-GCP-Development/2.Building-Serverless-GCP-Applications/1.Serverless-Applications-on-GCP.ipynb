# Introduction: Serverless and Cloud Functions

Welcome to the first lesson of the **Building Serverless Applications** course! In this lesson, we will explore the basics of serverless computing and how **Cloud Functions** help you build applications without managing infrastructure.

### What is Serverless?
Serverless computing means you do not have to worry about setting up or maintaining physical or virtual servers. Instead, you write your code, and the cloud provider (like Google Cloud) runs it for you only when needed. 

**Benefits include:**
* **No Maintenance:** No OS updates or hardware management.
* **Auto-Scaling:** The cloud automatically handles more users.
* **Cost Efficiency:** You only pay for the exact time your code is running.



### Cloud Functions & API Gateway
Cloud Functions are small, single-purpose pieces of code that run in response to events. In this lesson, we will connect a Cloud Function to an **API Gateway**, which acts as a "front door" for your application, passing user requests to your function for processing.

## Complete Code Example (`main.py`)

```python
import json

def calculate_tax(request):
    try:
        # Parse the input
        data = request.get_json(silent=True) or {}
        amount = float(data.get("amount", 0))
        
        # Perform calculations
        tax = round(amount * 0.07, 2)
        total = round(amount + tax, 2)
        
        # Prepare response
        response = {
            "amount": amount,
            "tax": tax,
            "total": total
        }
        
        return (json.dumps(response), 200, {"Content-Type": "application/json"})
        
    except Exception as e:
        # Handle errors
        return (json.dumps({"error": str(e)}), 400, {"Content-Type": "application/json"})
```

### Example Output
If a user sends a request with `{"amount": 100}`, the API returns:
```json
{
  "amount": 100.0,
  "tax": 7.0,
  "total": 107.0
}
```

## Fix the JSON Response Format in a Google Cloud Function

Great job on your first Cloud Function modification! Now let's work on a common best practice when building production APIs with Cloud Functions.

You have a Cloud Function that calculates tax correctly, but the response handling could be more robust. While returning a Python dictionary works in many cases, explicitly controlling JSON serialization gives you better consistency and makes debugging easier when things go wrong.

Your objective is to update the calculate_tax function to use explicit JSON serialization with json.dumps() instead of relying on automatic conversion. This approach ensures your response format is always predictable and follows API best practices.

Run the code first to see it working, then modify the response to use json.dumps() for the success case, matching the pattern already used in the error handler.

```python
import json

def calculate_tax(request):
    try:
        data = request.get_json(silent=True) or {}
        amount = float(data.get("amount", 0))
        tax = round(amount * 0.07, 2)
        total = round(amount + tax, 2)
        response = {
            "amount": amount,
            "tax": tax,
            "total": total
        }
        return (response, 200, {"Content-Type": "application/json"})
    except Exception as e:
        return (json.dumps({"error": str(e)}), 400, {"Content-Type": "application/json"})

```

When building APIs with Google Cloud Functions, explicitly serializing your data using `json.dumps()` is a best practice. It ensures that your response is a clean string and prevents unexpected behavior if the function attempts to return complex Python objects that the underlying HTTP server might not know how to handle.

By using `json.dumps()`, you gain full control over the string representation of your data, including indentation and character encoding.

Here is the corrected code with explicit JSON serialization for the success response:

```python
import json

def calculate_tax(request):
    try:
        data = request.get_json(silent=True) or {}
        amount = float(data.get("amount", 0))
        tax = round(amount * 0.07, 2)
        total = round(amount + tax, 2)
        
        response = {
            "amount": amount,
            "tax": tax,
            "total": total
        }
        
        # Explicitly serialize the response dictionary to a JSON string
        return (json.dumps(response), 200, {"Content-Type": "application/json"})
        
    except Exception as e:
        # The error handler already uses explicit serialization
        return (json.dumps({"error": str(e)}), 400, {"Content-Type": "application/json"})
```

### Why this is a "Production-Ready" move:

* **Consistency:** Now both your success path and your error path return the same data type (a string).
* **Header Alignment:** Since you are manually setting the `Content-Type` to `application/json`, it is safer to provide the body as a pre-formatted JSON string.
* **Predictability:** If your `response` dictionary ever grows to include complex types like `datetime` objects, `json.dumps()` allows you to define custom encoders to handle those specifically, rather than letting the framework guess.

## Add tax_rate Field to Cloud Function API Response

Excellent work debugging that JSON response issue! Now you're ready to enhance your API by making it more informative for clients who use your function.

Currently, your Cloud Function returns the amount, tax, and total, but it doesn't tell the client what tax rate was used in the calculation. This information could be valuable for users who want to understand how the tax was calculated or display it in their applications.

Your objective is to add a new field called tax_rate to the API response that contains the decimal value representing the tax rate used (which is currently 0.07 for 7%). Look for the TODO comment in the main.py file to see exactly where you need to make this change.

After completing this enhancement, when someone sends a request with an amount of 100, the response will include not just the calculated values, but also show them that a 7% tax rate was applied. This makes your API more transparent and helps clients understand exactly how their calculations

```python
import json

def calculate_tax(request):
    try:
        data = request.get_json(silent=True) or {}
        amount = float(data.get("amount", 0))
        tax = round(amount * 0.07, 2)
        total = round(amount + tax, 2)
        # TODO: Add "tax_rate": 0.07 to the response dictionary below
        response = {
            "amount": amount,
            "tax": tax,
            "total": total
        }
        return (json.dumps(response), 200, {"Content-Type": "application/json"})
    except Exception as e:
        return (json.dumps({"error": str(e)}), 400, {"Content-Type": "application/json"})

```

Adding metadata like the `tax_rate` to your response is a hallmark of good API design. It follows the principle of **self-documentation**, where the API provides enough context that a client doesn't have to guess how a calculation was performed.



Here is the updated code with the `tax_rate` field included in the response dictionary:

```python
import json

def calculate_tax(request):
    try:
        data = request.get_json(silent=True) or {}
        amount = float(data.get("amount", 0))
        
        # We use 0.07 as our constant tax rate
        current_rate = 0.07
        tax = round(amount * current_rate, 2)
        total = round(amount + tax, 2)
        
        # TODO: Add "tax_rate": 0.07 to the response dictionary below
        response = {
            "amount": amount,
            "tax_rate": current_rate,
            "tax": tax,
            "total": total
        }
        
        return (json.dumps(response), 200, {"Content-Type": "application/json"})
    except Exception as e:
        return (json.dumps({"error": str(e)}), 400, {"Content-Type": "application/json"})
```

### Why this improves your API:

* **Transparency:** If the tax rate changes in the future (e.g., to 8.5%), the client application can automatically update its UI to show the new rate without you needing to coordinate a front-end code change.
* **Auditability:** It provides a clear trace of the logic used at the exact moment the request was processed.
* **Flexibility:** In a more advanced version, you could eventually allow the client to pass in a `zip_code` and have the function look up and return a specific rate, while keeping the response structure identical.

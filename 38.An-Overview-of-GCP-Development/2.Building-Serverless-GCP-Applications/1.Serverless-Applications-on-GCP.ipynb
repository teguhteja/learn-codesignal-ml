# Introduction: Serverless and Cloud Functions

Welcome to the first lesson of the **Building Serverless Applications** course! In this lesson, we will explore the basics of serverless computing and how **Cloud Functions** help you build applications without managing infrastructure.

### What is Serverless?
Serverless computing means you do not have to worry about setting up or maintaining physical or virtual servers. Instead, you write your code, and the cloud provider (like Google Cloud) runs it for you only when needed. 

**Benefits include:**
* **No Maintenance:** No OS updates or hardware management.
* **Auto-Scaling:** The cloud automatically handles more users.
* **Cost Efficiency:** You only pay for the exact time your code is running.



### Cloud Functions & API Gateway
Cloud Functions are small, single-purpose pieces of code that run in response to events. In this lesson, we will connect a Cloud Function to an **API Gateway**, which acts as a "front door" for your application, passing user requests to your function for processing.

## Complete Code Example (`main.py`)

```python
import json

def calculate_tax(request):
    try:
        # Parse the input
        data = request.get_json(silent=True) or {}
        amount = float(data.get("amount", 0))
        
        # Perform calculations
        tax = round(amount * 0.07, 2)
        total = round(amount + tax, 2)
        
        # Prepare response
        response = {
            "amount": amount,
            "tax": tax,
            "total": total
        }
        
        return (json.dumps(response), 200, {"Content-Type": "application/json"})
        
    except Exception as e:
        # Handle errors
        return (json.dumps({"error": str(e)}), 400, {"Content-Type": "application/json"})
```

## Fix the JSON Response Format in a Google Cloud Function

Great job on your first Cloud Function modification! Now let's work on a common best practice when building production APIs with Cloud Functions.

You have a Cloud Function that calculates tax correctly, but the response handling could be more robust. While returning a Python dictionary works in many cases, explicitly controlling JSON serialization gives you better consistency and makes debugging easier when things go wrong.

Your objective is to update the calculate_tax function to use explicit JSON serialization with json.dumps() instead of relying on automatic conversion. This approach ensures your response format is always predictable and follows API best practices.

Run the code first to see it working, then modify the response to use json.dumps() for the success case, matching the pattern already used in the error handler.

```python
import json

def calculate_tax(request):
    try:
        data = request.get_json(silent=True) or {}
        amount = float(data.get("amount", 0))
        tax = round(amount * 0.07, 2)
        total = round(amount + tax, 2)
        response = {
            "amount": amount,
            "tax": tax,
            "total": total
        }
        return (response, 200, {"Content-Type": "application/json"})
    except Exception as e:
        return (json.dumps({"error": str(e)}), 400, {"Content-Type": "application/json"})

```

When building APIs with Google Cloud Functions, explicitly serializing your data using `json.dumps()` is a best practice. It ensures that your response is a clean string and prevents unexpected behavior if the function attempts to return complex Python objects that the underlying HTTP server might not know how to handle.

By using `json.dumps()`, you gain full control over the string representation of your data, including indentation and character encoding.

Here is the corrected code with explicit JSON serialization for the success response:

```python
import json

def calculate_tax(request):
    try:
        data = request.get_json(silent=True) or {}
        amount = float(data.get("amount", 0))
        tax = round(amount * 0.07, 2)
        total = round(amount + tax, 2)
        
        response = {
            "amount": amount,
            "tax": tax,
            "total": total
        }
        
        # Explicitly serialize the response dictionary to a JSON string
        return (json.dumps(response), 200, {"Content-Type": "application/json"})
        
    except Exception as e:
        # The error handler already uses explicit serialization
        return (json.dumps({"error": str(e)}), 400, {"Content-Type": "application/json"})
```

### Why this is a "Production-Ready" move:

* **Consistency:** Now both your success path and your error path return the same data type (a string).
* **Header Alignment:** Since you are manually setting the `Content-Type` to `application/json`, it is safer to provide the body as a pre-formatted JSON string.
* **Predictability:** If your `response` dictionary ever grows to include complex types like `datetime` objects, `json.dumps()` allows you to define custom encoders to handle those specifically, rather than letting the framework guess.

Excellent work on enhancing your API with additional response data! Now it's time to make your Cloud Function production-ready by adding proper error handling for invalid inputs.

Currently, your function assumes all requests will contain valid data, but real-world APIs need to handle bad inputs gracefully. Your function might crash or return incorrect results when users send missing data, negative numbers, or text instead of numbers.

Your objective is to add validation checks that protect your API from invalid requests. You need to:

Look for the TODO comments in the code to guide you on where to add these validation checks. When validation fails, your function should return helpful error messages that tell clients exactly what went wrong.

This exercise will teach you essential input validation patterns that make serverless APIs reliable and user-friendly in production environments.

```python
import json

def calculate_tax(request):
    try:
        data = request.get_json(silent=True) or {}
        
        # TODO: Check if "amount" field exists in the request data
        
        # TODO: Try to convert amount to float and handle conversion errors
        amount = float(data["amount"])
        
        # TODO: Check if amount is positive (greater than 0)
        
        # Calculate tax and total
        tax = round(amount * 0.07, 2)
        total = round(amount + tax, 2)
        
        response = {
            "amount": amount,
            "tax": tax,
            "total": total
        }
        return (json.dumps(response), 200, {"Content-Type": "application/json"})
        
    except Exception as e:
        return (json.dumps({"error": str(e)}), 400, {"Content-Type": "application/json"})

```

To make your Cloud Function truly "production-ready," you must move from an optimistic approach (assuming data is perfect) to a defensive one. Input validation acts as a filter that prevents your business logic from running on garbage data, which could otherwise lead to crashes or nonsensical financial calculations.



Here is the updated code with comprehensive validation checks:

```python
import json

def calculate_tax(request):
    try:
        data = request.get_json(silent=True) or {}
        
        # TODO: Check if "amount" field exists in the request data
        if "amount" not in data:
            return (json.dumps({"error": "Missing required field: 'amount'"}), 400, {"Content-Type": "application/json"})
        
        # TODO: Try to convert amount to float and handle conversion errors
        try:
            amount = float(data["amount"])
        except (ValueError, TypeError):
            return (json.dumps({"error": "'amount' must be a valid number"}), 400, {"Content-Type": "application/json"})
        
        # TODO: Check if amount is positive (greater than 0)
        if amount <= 0:
            return (json.dumps({"error": "'amount' must be greater than zero"}), 400, {"Content-Type": "application/json"})
        
        # Calculate tax and total
        tax = round(amount * 0.07, 2)
        total = round(amount + tax, 2)
        
        response = {
            "amount": amount,
            "tax": tax,
            "total": total
        }
        return (json.dumps(response), 200, {"Content-Type": "application/json"})
        
    except Exception as e:
        # General catch-all for unexpected errors
        return (json.dumps({"error": str(e)}), 400, {"Content-Type": "application/json"})
```

### Why these checks are vital:

* **Existence Check (`if "amount" not in data`):** Prevents a `KeyError` that would otherwise trigger the general `except` block. It allows you to give the user a specific, helpful message.
* **Type Validation (`ValueError/TypeError`):** If a user sends `{"amount": "abc"}`, the code will now gracefully explain that it needs a number rather than just crashing with a cryptic Python error.
* **Business Logic Check (`amount <= 0`):** In a financial API, processing a negative purchase amount could lead to your system "paying" the customer or causing accounting errors. Validating the range ensures your math always makes sense.

By returning a **400 Bad Request** status code for these failures, you correctly inform the client that the mistake was on their end (invalid input) rather than a bug in your function's code.

Here is your transformed `calculate_tax` function:

```python
import json

def calculate_tax(request):
    try:
        data = request.get_json(silent=True) or {}
        
        # Extract amount
        amount = float(data.get("amount", 0))
        
        # TODO: Extract tax_rate from the request body with a default of 0.07 (7%)
        tax_rate = float(data.get("tax_rate", 0.07))
        
        # TODO: Calculate tax instead of tip using amount * tax_rate
        tax = round(amount * tax_rate, 2)
        
        # TODO: Calculate total_with_tax instead of total
        total_with_tax = round(amount + tax, 2)
        
        # TODO: Update the response fields to include tax_rate, tax, and total_with_tax
        response = {
            "amount": amount,
            "tax_rate": tax_rate,
            "tax": tax,
            "total_with_tax": total_with_tax
        }
        
        return (json.dumps(response), 200, {"Content-Type": "application/json"})
        
    except Exception as e:
        return (json.dumps({"error": str(e)}), 400, {"Content-Type": "application/json"})
```

### Key Improvements in this Version:

* **Dynamic Flexibility:** By using `data.get("tax_rate", 0.07)`, you’ve implemented "Optional Parameters." If a user is in a state with a 5% tax, they can send `{"amount": 100, "tax_rate": 0.05}`. If they send nothing, your function gracefully falls back to the 7% default.
* **Logical Refactoring:** You've successfully pivoted the business logic from a service-industry "Tip" model to a retail "Tax" model by simply renaming variables and adjusting the math. 
* **Semantic Naming:** Changing the response key from `total` to `total_with_tax` provides more clarity to the person using your API, reducing the chance they'll be confused about whether tax was already included.

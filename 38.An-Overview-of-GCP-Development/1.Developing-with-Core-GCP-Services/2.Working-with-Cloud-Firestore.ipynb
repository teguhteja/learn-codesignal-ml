## Creating a Collection in Cloud Firestore

In Cloud Firestore, data is organized into **Collections**. 
* Each **Collection** contains **Documents**.
* Each **Document** holds a set of **Key-Value pairs** (similar to a JSON object).



> **Note:** Collections are created automatically by GCP the moment you add your first document to them. You don't need to "pre-create" them like you would a table in a SQL database.

### Example: Adding Your First Document
To create a unique collection and add a document, we use the following Python approach:

```python
import uuid
from google.cloud import firestore

# Initialize the Firestore Client
db = firestore.Client()

# Generate a unique collection name to avoid conflicts
COLLECTION_NAME = f"users_{uuid.uuid4().hex[:8]}"

# Create a reference to a document with a specific ID ("u-123")
doc_ref = db.collection(COLLECTION_NAME).document("u-123")

# .set() creates the document with the provided dictionary data
doc_ref.set({
    "name": "Ada",
    "email": "ada@example.com"
})

print(f"Created collection and added first document: {COLLECTION_NAME}")
```

### Breakdown of the Code:
* **`uuid` library:** Generates universally unique identifiers. Using `uuid.uuid4().hex[:8]` creates a short random suffix (like `1a2b3c4d`) so your collection name is unique.
* **`f-string`:** Combines the base name `users_` with the random suffix.
* **`.document("u-123")`:** Creates a reference to a document with a specific, human-readable ID. If you omit the ID, Firestore will generate a random one for you.
* **`.set({...})`:** Writes the data to the database.

**Sample Output:**
```text
Created collection and added first document: users_1a2b3c4d
```

## Creating a Collection in Cloud Firestore

## Performing CRUD Operations in Cloud Firestore

Now that you have a collection, let’s walk through the four basic operations.

### 1. Create: Adding a New Document
To add another document, use the `set` method with a new document reference:

```python
doc_ref = db.collection(COLLECTION_NAME).document("u-456")
doc_ref.set({
    "name": "Grace",
    "email": "grace@example.com"
})
print("Added user u-456")
```

## Cleaning Up: Deleting the Collection

Cloud Firestore does not provide a single method to delete an entire collection. Instead, you must delete every document within it. Here is a recursive function to handle the cleanup:

```python
def delete_collection(coll_ref, batch_size=10):
    docs = coll_ref.limit(batch_size).stream()
    deleted = 0
    
    for doc in docs:
        doc.reference.delete()
        deleted += 1
        
    if deleted >= batch_size:
        return delete_collection(coll_ref, batch_size)

delete_collection(db.collection(COLLECTION_NAME))
print("Deleted collection:", COLLECTION_NAME)
```

**Sample Output:**
> `Deleted collection: users_1a2b3c4d`

The reason your current code is failing to "update" correctly is that you are using `.set()` instead of `.update()`. 

In Firestore, **`.set()`** is a destructive operation by default—it overwrites the entire document with whatever new dictionary you provide. If you run your original code, the user's **"name"** field would actually disappear! To modify only specific fields while leaving the rest of the document intact, you must use the **`.update()`** method.

Here is the corrected code:

```python
import uuid
from google.cloud import firestore

db = firestore.Client()
COLLECTION_NAME = f"users_{uuid.uuid4().hex[:8]}"

def main():
    # Create collection and add first document
    doc_ref = db.collection(COLLECTION_NAME).document("u-123")
    doc_ref.set({
        "name": "Ada",
        "email": "ada@example.com"
    })
    print("Created collection and added first document:", COLLECTION_NAME)

    # Read document
    doc = doc_ref.get()
    print("Read:", doc.to_dict())

    # Update document
    # TODO: Fix the update operation so it properly updates the email field
    # Changing .set() to .update() ensures the "name" field is preserved.
    doc_ref.update({
        "email": "ada.lovelace@example.com"
    })
    
    updated_doc = doc_ref.get()
    print("Updated:", updated_doc.to_dict())

    # Cleanup: delete collection
    def delete_collection(coll_ref, batch_size=10):
        docs = coll_ref.limit(batch_size).stream()
        deleted = 0
        for doc in docs:
            doc.reference.delete()
            deleted += 1
        if deleted >= batch_size:
            return delete_collection(coll_ref, batch_size)
    delete_collection(db.collection(COLLECTION_NAME))
    print("Deleted collection:", COLLECTION_NAME)

if __name__ == "__main__":
    main()
```

### Why this fix is critical:

* **Data Integrity:** If you had a document with 20 fields and only wanted to change the password, using `.set()` without the `merge=True` flag would delete the other 19 fields.
* **Method Purpose:** `.update()` explicitly tells Firestore: "Find this document and only change these specific keys." If the document doesn't exist, `.update()` will actually throw an error, which is a helpful safety check to ensure you aren't creating data where you didn't intend to.

**Quick Tip:** If you ever *do* want to use `.set()` but keep existing fields, you can use `doc_ref.set({"email": "..."}, merge=True)`. However, for simple modifications, `.update()` is the standard and most readable choice!

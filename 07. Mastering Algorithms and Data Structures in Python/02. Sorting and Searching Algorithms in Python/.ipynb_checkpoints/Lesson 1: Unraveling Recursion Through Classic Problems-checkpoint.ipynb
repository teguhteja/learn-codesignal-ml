## Problem 1: Fibonacci Sequence

### What is the Fibonacci Sequence?
The Fibonacci sequence is an infinite sequence of numbers that starts with 0 and 1. Each subsequent number is the sum of the previous two:

`0, 1, 1, 2, 3, 5, 8, 13, ...`

Given a number `n`, our task is to implement a recursive algorithm that returns the n-th number in the Fibonacci sequence. Remember, the indexing starts from 0.

> ðŸ’¡ **Fun Fact:** The Fibonacci sequence is closely connected with the golden ratio, often found in nature and art.

---

### Naive Recursion Approach

Hereâ€™s a simple recursive solution to find the n-th Fibonacci number:

```python
def fib(n): 
   if n <= 1: 
       return n 
   else: 
       return fib(n - 1) + fib(n - 2)
```

While this solution works, it is inefficient due to repeated calculations. For example, when calculating the 5th Fibonacci number, the 3rd Fibonacci number will be calculated multiple times. This leads to exponential time complexity `O(2^n)`.

---

### Efficient Approach with Memoization

To optimize, we use **memoization**â€”storing the results of previous calculations to avoid redundancy.

```python
def fib(n, computed={0: 0, 1: 1}):
    if n not in computed:
        computed[n] = fib(n - 1, computed) + fib(n - 2, computed)
    return computed[n]
```

This version computes the Fibonacci number in linear time `O(n)`, which is a huge improvement over the naive recursive solution.

## Problem 2: Summing Array Elements

Our next task is to calculate the sum of all elements in an array. Although it might seem trivial, recursion provides an elegant way to solve this.

### Recursive Approach Explanation

We decompose the array sum problem by breaking it into two parts: the first element and the sum of the remaining elements.

```python
def arraySum(arr, index=0): 
   if index == len(arr): 
       return 0 
   else:
       return arr[index] + arraySum(arr, index + 1)
```

This approach uses an index to walk through the array, efficiently summing the elements without creating new arrays. The base case occurs when the index reaches the array length.

## Problem 3: Calculating Factorial

The factorial of a number `n` is the product of all positive integers less than or equal to `n`:

```
n! = n * (n-1) * (n-2) * ... * 2 * 1
```

### Recursive Approach Explanation

The factorial of `n` can also be expressed recursively as:

```
n! = n * (n-1)!
```

### Solution:

```python
def factorial(n): 
   if n == 0 or n == 1: 
       return 1
   else:
       return n * factorial(n - 1)
```

For values of `n = 0` or `n = 1`, the function returns 1. For larger `n`, the function recursively computes the factorial.

### Plan:
1. **Base Case**: If the index exceeds the length of the array, return 0 (this handles the edge case of an empty array).
2. **Recursive Case**: If the index is even, add the current element to the sum of the treasures from the next even index (index + 2). If the index is odd, simply move to the next index (index + 1).

Hereâ€™s how we can implement this:

```python
def recursiveSumEven(arr, idx=0):
    # Base case: if the index is out of bounds, return 0
    if idx >= len(arr):
        return 0
    # If the index is even, include the current element in the sum
    if idx % 2 == 0:
        return arr[idx] + recursiveSumEven(arr, idx + 2)
    else:
        # If the index is odd, skip to the next index
        return recursiveSumEven(arr, idx + 1)

# Testing the function
print(recursiveSumEven([1, 2, 3, 4, 5, 6]))  # Expected output: 9
print(recursiveSumEven([2, 3]))               # Expected output: 2
print(recursiveSumEven([]))                    # Expected output: 0
```

### Explanation:
- **Base Case**: The function checks if the current index `idx` is greater than or equal to the length of the array. If so, it returns 0, indicating there are no more elements to sum.
- **Even Index Handling**: If the index is even, it adds the value at that index to the result of the recursive call for the next even index (`idx + 2`).
- **Odd Index Handling**: If the index is odd, it simply calls itself for the next index (`idx + 1`), effectively skipping the current odd-indexed element.

### Plan:
1. **Factorial Function**: This function will compute the factorial of a given number using recursion and memoization to optimize repeated calculations.
2. **Factorials Function**: This function will iterate through the list of numbers, applying the factorial function to each and returning a new list with the results.

Hereâ€™s how we can implement this:

```python
def factorial(num, memo={}):
    # Handle negative numbers
    if num < 0:
        return None  # Factorial is not defined for negative numbers
    # Base case for 0! = 1
    if num == 0:
        return 1
    # Check if the factorial has already been computed
    if num in memo:
        return memo[num]
    # Recursive calculation
    memo[num] = num * factorial(num - 1, memo)
    return memo[num]

def factorials(nums):
    return [factorial(num) if factorial(num) is not None else 'Error' for num in nums]

# Testing the function
print(factorials([2, 3, 4]))        # should print: [2, 6, 24]
print(factorials([1, 5, 6]))        # should print: [1, 120, 720]
print(factorials([0, -3, 10]))      # should print: [1, 'Error', 3628800]
```

### Explanation:
- **Factorial Function**:
  - **Negative Check**: If the input number is negative, it returns `None`, indicating an error.
  - **Base Case**: If the number is 0, it returns 1 (since 0! = 1).
  - **Memoization**: It checks if the factorial of the number has already been computed and stored in the `memo` dictionary. If so, it returns the stored value.
  - **Recursive Calculation**: If not, it calculates the factorial recursively and stores the result in `memo` for future reference.

- **Factorials Function**:
  - It uses a list comprehension to apply the `factorial` function to each number in the input list. If the factorial function returns `None`, it adds 'Error' to the result list.

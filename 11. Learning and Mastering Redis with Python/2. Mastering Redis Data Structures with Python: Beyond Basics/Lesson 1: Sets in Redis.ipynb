This example shows how to handle sets in Redis and how simple it is to perform operations on them.

### Let's break down the code:

1. We first import the `redis` module and connect to the Redis server.
2. We then add items to a set called `countries` using the `sadd` command.
3. Finally, we retrieve all members of the set using the `smembers` command and print them out. The result will be `Countries in the set: ['USA', 'Canada', 'UK']` - notice that the duplicate 'USA' was not added to the set. Also, keep in mind that the order of the elements in the set is not guaranteed.

Let's familiarize ourselves with the basic operations on sets in Redis. Particularly, we will learn how to get the number of items in a set and remove an item from a set:

```python
# Get the number of items in the set
num_countries = client.scard('countries')
print(f"Number of countries in the set: {num_countries}") # Output: Number of countries in the set: 3

# Remove an item from the set
client.srem('countries', 'UK') # Remove 'UK' from the set
```

In this code snippet, we use the `scard` command to get the number of items in the set and the `srem` command to remove an item from the set.

Here's a Python script that utilizes Redis to add unique movie titles to a set named `movies`, retrieves them, and prints the titles:

```python
import redis

# Connect to the Redis server
client = redis.Redis(host='localhost', port=6379, db=0)

# Add unique movie titles to the 'movies' set
movie_titles = ['Inception', 'The Matrix', 'Interstellar', 'Inception']
for title in movie_titles:
    client.sadd('movies', title)

# Print the size of the 'movies' set
movies_count = client.scard('movies')
print('Number of unique movies:', movies_count)

# Retrieve all movie titles from the 'movies' set
unique_movies = client.smembers('movies')

# Decode the bytes and print the list of movie titles
decoded_movies = [movie.decode('utf-8') for movie in unique_movies]
print('Unique movie titles:', decoded_movies)
```

### Explanation:
1. **Import Redis**: The `redis` library is imported to work with Redis databases.
2. **Connect to Redis**: A Redis client is created, connecting to the local Redis server.
3. **Add Movie Titles**: The `sadd` method adds each title to the `movies` set. If a title already exists, it won't be added again (ensuring uniqueness).
4. **Count Unique Movies**: The `scard` method retrieves the number of unique entries in the set.
5. **Retrieve and Decode Titles**: The `smembers` method retrieves all unique movie titles as bytes, which are then decoded into strings using a list comprehension.
6. **Print Results**: Finally, it prints the count of unique movies and the list of titles.

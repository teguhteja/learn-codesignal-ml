## Why Projects Need Instructions

Before diving into code, let's understand the problem we're solving. Imagine working on a data analysis project where you have established specific technical standards:

* Using `pandas` for all internal data manipulation.
* Formatting all currency reporting strictly as USD.
* Following PEP 8 style guidelines for naming conventions.
* Saving data visualizations at a resolution of 300 DPI.

Without project instructions, you would need to remind Claude about these granular conventions in every single session. Even worse, Claude might make inconsistent architectural choices, using entirely different libraries, formatting styles, or syntax patterns each time you start a new thread.

Project instructions solve this by creating a **persistent context** that Claude reads at the very start of every session. This isn't a magical, hidden memory; rather, Claude loads the localized markdown file automatically and directly applies those specific design constraints to all its generative and tool-based work. The result is consistent, standardized source code that matches your project's exact conventions without repeated explanations.

## Creating Your First `CLAUDE.md`

Let's look at how to create project instructions using the `/init` command. When we enter `/init` in Claude Code, it runs a deep-scan analysis of our repository structure—examining key configuration files, project dependencies, and general folder organization to generate highly tailored instructions:

```text
> /init
● I'll analyze the codebase to create a comprehensive CLAUDE.md file. 
  Let me start by exploring the repository structure and key files.

● Bash(find /usercode/FILESYSTEM -maxdepth 2 -type f -name "*.md"...)
  ⎿ /usercode/FILESYSTEM/.codesignal/requirements.txt
    /usercode/FILESYSTEM/.codesignal/README.md

● Read(sales_data.csv)
  ⎿ Read 61 lines

● Read(.codesignal/requirements.txt)
  ⎿ Read 2 lines

● Write(CLAUDE.md)
  ⎿ Wrote 49 lines to CLAUDE.md

● I've created a CLAUDE.md file for this repository...

```

The `/init` command actively inspects our project by reading active environment configurations, evaluating dataset shapes, and identifying explicit dependencies. This localized analysis ensures that the generated `CLAUDE.md` matches our actual codebase layout rather than throwing a generic boilerplate template at us.

Notice how highly targeted these appended guidelines are. The `pandas` directives don't just say "write clean python"; they explicitly require date parsing rules, force descriptive naming styles, and dictate how data aggregations should be performed.

## How Claude Leverages Instructions

Let's see these rules actively shape behavior. Watch what happens when we ask Claude to construct a function to calculate monthly business revenue:

```text
> Create a function to calculate monthly revenue
● Based on your project standards, I'll use pandas and proper formatting...

● Write(revenue.py)
  ⎿ Wrote 12 lines to revenue.py

```

Claude explicitly calls out the `CLAUDE.md` files ruleset before generating a single line of text. The resulting output maps exactly to our requirements:

```python
import pandas as pd

def calculate_monthly_revenue(df: pd.DataFrame) -> pd.DataFrame:
    """Calculate total revenue grouped by month.
    
    Args:
        df: Sales data with 'date' and 'amount' columns
        
    Returns: 
        DataFrame with month and revenue_usd columns
    """
    df['month'] = pd.to_datetime(df['date']).dt.to_period('M')
    monthly = df.groupby('month')['amount'].sum()
    return monthly.to_frame('revenue_usd').round(2)

```

### Evaluating Rule Enforcement:

* **Pandas Focus:** The routine relies entirely on pandas primitives (`to_datetime`, `groupby`), following our requirement to avoid native looping or alternative arrays.
* **Documentation Rules:** The script generates a fully populated docstring tracking inputs and outcomes, directly satisfying our *Documentation Requirements*.
* **Schema Alignment:** The logic targets the exact fields (`date` and `amount`) specified in our data model layout.
* **Styling Best Practices:** The explicit variable naming conventions and currency rounding behaviors are perfectly adhered to out of the box.

## Architecture Layering: Team vs. Personal Instructions

Claude Code features a clever tiered system for loading instructions that perfectly respects the balance between collaborative team norms and custom developer preferences:

```text
┌───────────────────────────────┐
│     1. CLAUDE.md              │  ◄── Committed to Version Control (Git)
│     (Shared Team Standards)   │      Contains style guides, test paths, compile scripts.
└──────────────┬────────────────┘
               │
               ▼
┌───────────────────────────────┐
│     2. CLAUDE.local.md        │  ◄── Added to .gitignore
│     (Personal Dev Overlay)    │      Contains local DB paths, ports, custom paths.
└───────────────────────────────┘

```

### 1. `CLAUDE.md` (The Team Blueprint)

This file holds project-wide standards and workflow behaviors. It **must be committed to version control**. This guarantees every team member—and every Claude Code agent working on the repository—shares the same coding conventions, engine profiles, and building restrictions.

### 2. `CLAUDE.local.md` (The Developer Overlay)

This file holds personal preferences and credentials specific to your environment. It **should always be listed inside your `.gitignore**`. This is where you specify custom machine paths, developer-specific container endpoints, or internal debugging ports.

> 🛠️ **Layering Precedence:** When running workspace tasks, Claude parses `CLAUDE.md` first to build the foundational team constraints, and then instantly applies the `CLAUDE.local.md` guidelines as a localized override. This lets you personalize your workflow without creating merge conflicts in your team's main branch.

## Core Characteristics of Great Instructions

To ensure your instructions engine runs perfectly, construct your configuration profiles with these traits in mind:

* **Keep it Concise:** Focus strictly on operational guidance. Do not copy long articles or documentation into your file. If a rule does not directly influence Claude's tool choice or output format, leave it out.
* **Be Specific:** Avoid abstract statements like *"Write high-quality logic"*. Instead, use precise rules like *"Use descriptive variable names for DataFrames (e.g., sales_df)"*.
* **Provide Executable Commands:** Tell Claude exactly how to verify its changes. Instead of writing *"Test the application"*, add an explicit entry: *"Run bash main.sh to execute testing hooks"*.
* **Expose Schemas Explicitly:** If your app consumes flat files, APIs, or database records, map out the column headers, keys, and data formats directly within the instructions.
* **Keep it Organized:** Use clear section headers, clean markdown lists, and bullet points. Structuring data cleanly ensures Claude reads and applies your parameters without losing context.

## What Does "Agentic" Mean?

The term **agentic** refers to Claude Code's ability to act autonomously on our behalf rather than merely providing suggestions. When we describe a goal, Claude does not just tell us what to do; it actually does it by selecting and using the appropriate tools.

Think of the difference between a traditional coding assistant that tells us *"write this function here"* and Claude Code that actually writes the function. We provide the destination, and Claude takes the wheel, making decisions and performing actions to get us there.

This autonomy extends across the entire development workflow. Claude can create new files, read and analyze existing code, modify files with surgical precision, execute shell commands, and fetch information from the web. Each action uses a specific tool, and Claude decides which tools to employ based on our goals.

## Claude's Toolbox

Claude Code has access to several specialized tools organized into distinct functional areas:

### File Operations

* **Read:** Examines files (including source code, images, PDFs, and notebooks).
* **Write:** Creates brand-new files from scratch.
* **Edit:** Modifies existing files with precise, localized string replacements.
* **Glob:** Finds files across directories using structural patterns (e.g., `/*.ts`).
* **Grep:** Searches file contents matching regular expressions (regex).

### Code Execution

* **Bash:** Executes shell commands directly in your local environment (e.g., `git`, `npm`, `docker`).
* **BashOutput:** Monitors standard outputs of long-running background processes.
* **KillShell:** Safely terminates active background processes.

### Web & Search

* **WebFetch:** Fetches and analyzes remote web page data contents.
* **WebSearch:** Queries the web for current, up-to-date information.

### Additional Capabilities

* **Task Management:** Launches specialized internal sub-agents and tracks active task lists.
* **Jupyter Notebooks:** Directly reads and edits specialized data science notebook cells.
* **User Interaction:** Prompts you with clarifying questions or executes custom user commands.

For this lesson, we will focus primarily on the core **file operations** and **code execution** tools. You will encounter the others naturally as your projects grow more sophisticated.

## Creating Files with `Write`

Let's see these tools in action. Watch what happens when we ask Claude to create a simple Python file:

```text
> Create a Python file called hello.py that prints "Hello, World!"
● Write(hello.py)

──────────────────────────────────────────────────────────────────────────────────────────────────────────────
 Create file
╭────────────────────────────────────────────────────────────────────────────────────────────────────────────╮
│ hello.py                                                                                                   │
│                                                                                                            │
│ print("Hello, World!")                                                                                     │
╰────────────────────────────────────────────────────────────────────────────────────────────────────────────╯
 Do you want to create hello.py?
 ❯ 1. Yes
   2. Yes, allow all edits during this session (shift+tab)
   3. No, and tell Claude what to do differently (esc)

```

Several important operational details emerge here:

1. **Autonomous Selection:** Claude automatically chose the `Write` tool—we did not specify this mechanism.
2. **Pre-Flight Preview:** Before making any concrete changes to your filesystem, Claude shows us a clear preview of what will be created. This is the permission system in action.

### The Three Authorization Options:

* **`Yes`:** Approves this specific file operation.
* **`Yes, allow all edits during this session`:** Instructs the workspace to trust Claude to make edits across files without prompting again.
* **`No, and tell Claude what to do differently`:** Rejects the change and provides immediate feedback.

Once we select `Yes`, Claude proceeds to execute the instruction:

```text
● Write(hello.py)  ⎿  Wrote 2 lines to hello.py
     print("Hello, World!")

● Done! I've created hello.py with a simple print statement.

```

The indented output confirms exactly what was written and provides a friendly confirmation. The file now exists in our working directory, ready to run.

## Modifying Files with `Edit`

Creating new files is essential, but what about changing existing ones? This is where `Edit` shines. Unlike `Write`, which creates files completely from scratch, `Edit` makes targeted modifications to existing files without altering unchanged background content.

Let's modify our `hello.py` file:

```text
> Update hello.py to also print "Welcome to Claude Code!"
● I'll update the file to include both print statements.

● Read(hello.py)
  ⎿  Read 2 lines

● Edit(hello.py)

──────────────────────────────────────────────────────────────────────────────────────────────────────────────
 Edit file hello.py
╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌
   1    print("Hello, World!")
   2 +  print("Welcome to Claude Code!")
╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌⛌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌╌
 Do you want to make this edit to hello.py?
 ❯ 1. Yes
   2. Yes, allow all edits during this session (shift+tab)
   3. No, and tell Claude what to do differently (esc)

```

Notice how the tool sequences this operation behind the scenes:

1. **Context Check:** Claude automatically triggers `Read` on the file first to understand its existing layout.
2. **Diff Generation:** It loads the `Edit` tool. The interface prints a clean unified diff format with line numbers: line 1 remains unchanged, and line 2 (marked with a **`+`**) is ready to be added.

This precision is crucial: Claude preserves existing content while editing exactly what we requested. After approving:

```text
● Edit(hello.py)  ⎿  Updated hello.py with 1 addition       
       1    print("Hello, World!")
       2 +  print("Welcome to Claude Code!")

● Done! The file now prints both messages. When you run python hello.py, it will output:
  Hello, World!
  Welcome to Claude Code!

```

### The Difference at a Glance

* **`Write`:** Used when you say *"create a new test module"*. It maps out an entirely new file buffer from scratch.
* **`Edit`:** Used when you say *"add error handling to our login function"*. It targets a localized block to swap or append code rows dynamically.

## Executing Commands with `Bash`

Creating and modifying files is only half the story. Let's execute our Python script to see `Bash` in action:

```text
> Run it
● Bash(python hello.py)
  ⎿  Running…

──────────────────────────────────────────────────────────────────────────────────────────────────────────────
 Bash command

   python hello.py
   Run hello.py script

 Do you want to proceed?
 ❯ 1. Yes
   2. Yes, and don't ask again for python commands in /usercode/FILESYSTEM
   3. Tell Claude what to do differently

 Esc to exit

```

Claude selected the `Bash` tool and derived the correct command implicitly from context. We simply said *"Run it"*, and Claude translated that intent into the exact terminal invocation.

Notice that the `Bash` execution pipeline also adheres to our permission layout—showing you exactly what will run before requesting permission. After approving:

```text
● Bash(python hello.py)  ⎿  Hello, World!          
     Welcome to Claude Code!

● Perfect! The script ran successfully and printed both messages.

```

The execution results return indented cleanly underneath the tool invocation block, showing the output results directly inside the session stream.

## Tool Chaining in Action

Often, Claude chains multiple tools sequentially to resolve an abstract goal. This coordination happens automatically behind the scenes. For instance, if you ask Claude to *"fix the typo in config.py where 'databse' should be 'database'"*, the tool chain responds like this:

```text
┌──────────────┐      ┌──────────────┐      ┌──────────────┐
│  1. Read     ├─────►│  2. Edit     ├─────►│  3. Bash     │
│  Examine file│      │  Patch typo  │      │  Run tests & │
│  to find bug │      │  with string │      │  verify compile
└──────────────┘      └──────────────┘      └──────────────┘

```

You do not need to break tasks down into micro-steps. You focus on describing the macro target outcome, and Claude orchestrates the tool sequence required to get there.

## The Permission System Guardrails

The permission system provides complete control over Claude's actions. Claude balances autonomy with absolute human oversight, keeping you in the driver's seat while taking care of the mechanical details of execution.

### Tool Authorization Default Policies

* **File Mutations (`Write`, `Edit`):** Always pause to request interactive human verification by default.
* **File Introspection (`Read`, `Glob`, `Grep`):** Typically execute silently without prompting to streamline speed.
* **Command Executions (`Bash`):** Prompt for explicit confirmation depending on the environmental risk profile of the instruction.

### The Three-Option Prompt Actions

1. **`Yes`:** Approves only this single, isolated operation. Claude will pause to ask again on the very next file mutation. Ideal when reviewing major changes carefully.
2. **`Yes, allow all edits during this session`:** Switches to a streamlined workflow. Claude will make file modifications without interrupting you for the remainder of your workspace session. You still see every log line and code block flash by, but you don't need to manually click approval.
3. **`No, and tell Claude what to do differently`:** Completely rejects the operation and opens a text field to pass corrective guidance. Claude will rewrite its strategy based on your feedback.

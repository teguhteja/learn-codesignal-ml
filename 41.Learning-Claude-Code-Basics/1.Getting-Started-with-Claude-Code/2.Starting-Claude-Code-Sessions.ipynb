## Working in the Terminal

Before we launch Claude Code, let's take a moment to understand where we will be working. Claude Code operates entirely within the terminal, also called the command line or shell. This is the text-based interface where we type commands and see results.

If you are new to working in terminals, do not worry. Using Claude Code does not require advanced terminal skills. The terminal is simply where we start Claude Code and where it displays its messages. Think of it as opening an application, but instead of clicking an icon, we type a command.

* **Windows:** Users can leverage **Command Prompt**, **PowerShell**, or the **Windows Subsystem for Linux (WSL)**.

## Understanding Session Management

As we work with Claude Code, we will naturally accumulate a comprehensive conversation history: messages we have sent, responses from Claude Code, code that has been written, and commands that have been executed. This accumulated history forms a **Session**.

Sessions are critically important because Claude Code uses the entire conversation history as continuous context for understanding what we are trying to accomplish. If we are building a complex feature across multiple distinct messages, Claude Code remembers previous decisions and stays consistent with earlier work.

However, there are times when we need to actively manage these sessions: starting fresh when switching tasks, closing Claude Code when we are finished, or returning to previous work. Claude Code provides native slash-commands for each of these scenarios.

---

## Core Session Management Commands

### 1. Starting Fresh with `/clear`

Sometimes we want to wipe the slate clean and begin a new conversation within the same session. Perhaps we have finished one task and want to start a completely different one, or maybe the conversation has become cluttered with exploratory attempts that are no longer relevant to the core objective.

The `/clear` command removes all active conversation history from the context layer and gives us a clean starting point:

```text
> /clear
⎿ (no content)

> 

```

After clearing, Claude Code shows `⎿ (no content)`, indicating that the active conversation history has been safely removed. The prompt returns immediately, ready for your next message.

Importantly, `/clear` **does not exit** Claude Code; we are still running in the same terminal session, just with an empty conversation memory frame. This is incredibly useful when we want to context-switch between completely unrelated tasks without leaving the tool entirely.

### 2. Exiting Cleanly with `/exit`

When we are done working and want to close Claude Code completely, we use the `/exit` command. This terminates the process loop and safely returns us back to our operating system's native terminal prompt:

```text
> /exit
⎿ See ya!
/$ 

```

Claude Code displays a friendly farewell message and then shuts down. Notice how the prompt indicator changes from the internal active **`>`** indicator back to your standard system shell signature **`/$`**. We are no longer in a Claude Code session.

Exiting is what we do at the end of a work session when we are truly finished and want to close the asset. Unlike `/clear`, which just resets the short-term conversation thread, `/exit` closes the Claude Code process entirely.

### 3. Returning to Previous Work with `/resume`

One of the most practical features of Claude Code is its ability to automatically save your conversation history to disk. When we exit, our session state is not lost; it is stored safely and can be resumed later. This means we can pick up exactly where we left off, even after closing our terminal windows or restarting our computer.

To return to a previous conversation context, launch Claude Code again and enter the `/resume` operator:

```text
❯ /resume ───────────────────────────────────────────────────────────────────────────────────────────────────────────────────────────────────

Resume Session
╭─────────────────────────────────────────────────────────────────────────────────────────────────────────────────────────────────╮
│ ⌕ Search…                                                                                                                       │
╰─────────────────────────────────────────────────────────────────────────────────────────────────────────────────────────────────╯

❯ Hi
  29 seconds ago · 4 messages · -

  A to show all projects ·  · P to preview · R to rename · Type to search · escape to cancel · 

```

The `/resume` command presents a dynamic list of our recent sessions. Each explicit log entry displays:

* The session name (auto-generated based on the initial message content).
* A relative timestamp tracking how long ago we worked on it.
* The explicit message count metrics contained within the thread history.
* The workspace directory path where the work took place.

We can navigate this menu to select which session to continue. This is helpful when we are juggling multiple projects or need to quickly resume work started yesterday.

## Choosing the Right Operator

Now that we have explored `/clear`, `/exit`, and `/resume`, let's clarify exactly when to implement each command to keep our software delivery workflow running smoothly.

| Command Scenario | Best Used When... | Operational Result |
| --- | --- | --- |
| **`/clear`** | You want to start a completely new programming task but stay inside the tool's runtime loop. | The terminal session continues uninterrupted, but the volatile conversation history is erased. This is like opening a new clean document inside a text editor without closing the application wrapper. |
| **`/exit`** | You are completely finished working for the block and want to close the AI assistant down. | The system terminates the active process entirely and returns control to your regular shell prompt. It's the standard teardown action to perform at the end of the day. |
| **`/resume`** | You have booted up Claude Code and want to jump directly back into an active codebase stream. | Re-establishes an archived session with its entire multi-turn context loop fully intact, ensuring structural consistency with earlier choices. |

Understanding these distinctions helps us manage our workflows efficiently. We avoid constantly starting from scratch when we don't need to, and we avoid clogging our context windows with unnecessary background logs when we want a completely clean slate.

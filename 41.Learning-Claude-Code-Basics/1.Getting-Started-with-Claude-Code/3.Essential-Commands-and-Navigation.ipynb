## Checking Session Status

The first and most fundamental command is `/status`, which provides a detailed snapshot of our current session. This command reveals the inner workings of our environment through a tabbed interface. Let's see what information it reveals:

```text
> /status 
──────────────────────────────────────────────────────────────────────────────────────
 Settings:  Status   Config   Usage   (tab to cycle)

 Version: 2.0.42
 Session ID: 3945885e-0866-48d6-b08a-825b2d504756
 cwd: /home/user/project
 Auth token: ...
 API key: ...

 Model: Default (claude-sonnet-4.5)
 Memory:
 Setting sources: User settings, Shared project settings, Local
──────────────────────────────────────────────────────────────────────────────────────

```

The output is organized into tabs that you can cycle through using the **Tab** key. In the **Status** view, we see several critical pieces of information:

* **Session ID:** A unique identifier for your current work.
* **cwd (Current Working Directory):** Shows exactly where Claude Code will create and modify files.
* **Model:** Confirms which AI version is currently processing your requests.

This command becomes invaluable when something seems off. If Claude Code is not finding files we know exist, checking the `cwd` confirms we are in the right directory. If you are unsure why Claude is behaving a certain way, the `Setting sources` line tells you which configuration files (like *Local* or *Shared project settings*) are influencing the session. The `/status` command is our first line of investigation.

## Understanding Model Options

Claude Code does not use just one AI model; instead, it offers three different models, each with distinct characteristics. These models sit along a spectrum of capability, speed, and cost. Understanding this spectrum helps us choose the right tool for each task.

The three models are **Haiku**, **Sonnet**, and **Opus**:

* **Haiku:** The fastest and most economical option, designed for quick tasks and rapid responses. It processes requests quickly but may struggle with intricate architectural decisions.
* **Sonnet:** The default and recommended model. It perfectly balances capability and cost for general development work, handling sophisticated reasoning for most development scenarios.
* **Opus:** Available as a legacy option (version 4.1) that may still be useful for certain complex tasks, though it comes with higher costs and may eventually be phased out.

Knowing these distinctions allows us to optimize our workflow efficiently.

## The Model Selector

To see which models are available and switch between them, we use the `/model` command. When we enter this command without arguments, Claude Code presents an interactive selector showing all available options:

```text
> /model ──────────────────────────────────────────────────────────────────────────────────────────────────────────────
 Select model
 Switch between Claude models. Applies to this session and future Claude Code sessions. For other/previous 
 model names, specify with --model.

 ❯ 1. Default (recommended)   Use the default model (currently Sonnet 4.5) · $3/$15 per Mtok ✔
   2. Opus                    Legacy: Opus 4.1 for complex tasks · $15/$75 per Mtok
   3. Haiku                   Haiku 4.5 for simple tasks · $1/$5 per Mtok

 Enter to confirm · Esc to exit

```

The selector displays each model with its pricing structure. The format **$X/$Y per Mtok** shows the cost per million tokens: the first number is for input tokens (what we send to Claude), and the second is for output tokens (what Claude generates).

Notice the cost breakdown:

* **Sonnet:** $3 for input and $15 for output.
* **Haiku:** Significantly cheaper at $1 input and $5 output.
* **Opus (Legacy):** The most expensive at $15 input and $75 output.

The arrow (**❯**) indicates which option is currently selected. We can navigate this menu using arrow keys to choose a different model, or we can press **Enter** to keep the current selection.

---

## Switching Models Directly

While the interactive selector works well, we can also switch models directly by specifying the model name as an argument. This is faster when we know exactly which model we want. Let's switch to Haiku:

```text
> /model haiku   ⎿  Set model to haiku (claude-haiku-4-5-20251001)

```

The confirmation message tells us the switch was successful and shows the specific model identifier. Now all subsequent interactions in this session will use Haiku until we switch again or start a new session.

This direct approach is particularly useful in workflows where we want a specific model for a particular phase of work. For example, we might use Haiku for quick edits and formatting, then switch to Sonnet when tackling more complex logic.

## Choosing the Right Model

Choosing the right model involves balancing three factors: **speed, capability, and cost**.

### Use Haiku

For routine tasks that do not require deep reasoning:

* Fixing typos or formatting code.
* Writing simple test cases.
* Generating boilerplate code.
* *Haiku excels at well-defined tasks where the approach is straightforward.*

### Use Sonnet (Default)

As your default choice for general development work:

* Implementing complete features.
* Refactoring existing codebases.
* Writing documentation or debugging moderate to high complexity issues.
* *Sonnet handles most development scenarios effectively, including complex challenges.*

### Use Opus (Legacy)

Sparingly and only if you have specific legacy workflows that rely on it:

* *Since it is marked as legacy, it may be deprecated in future versions. For most complex tasks that traditionally called for a premium model, Sonnet 4.5 is now the better, recommended choice.*

> 💰 **Cost Note:** The pricing difference is significant. Using Opus for simple tasks could cost **five times more** than Haiku. Matching the model to the task is both economical and effective.

## Inspecting Context

Another essential command is `/context`, which shows how Claude Code is using its available working memory. Think of this as checking how full Claude's "workspace" is during your session. Let's see what this command reveals:

```text
> /context   ⎿  
      Context Usage
     ⛁ ⛁ ⛁ ⛁ ⛁ ⛁ ⛁ ⛀ ⛀ ⛶   claude-sonnet-4-5-20250929 · 61k/200k tokens (30%)
     ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ 
     ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶   ⛁ System prompt: 2.4k tokens (1.2%)
     ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶   ⛁ System tools: 13.4k tokens (6.7%)
     ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶   ⛁ Messages: 8 tokens (0.0%)
     ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶   ⛶ Free space: 139k (69.6%)
     ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶   ⛝ Autocompact buffer: 45.0k tokens (22.5%)
     ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛶ ⛝ ⛝ ⛝ 
     ⛝ ⛝ ⛝ ⛝ ⛝ ⛝ ⛝ ⛝ ⛝ ⛝ 
     ⛝ ⛝ ⛝ ⛝ ⛝ ⛝ ⛝ ⛝ ⛝ ⛝ 

     SlashCommand Tool · 0 commands
     └ Total: 864 tokens

```

The most important line is at the very top: **`61k/200k tokens (30%)`**. This tells us we are using 61,000 out of 200,000 available tokens, or about 30% of Claude's working memory. The visual bar provides an at-a-glance view of this allocation.

### The Context Space Breakdown:

* **System Prompt & System Tools:** The baseline overhead that Claude Code requires to function.
* **Messages:** Our active conversation history, which in a fresh session is minimal.
* **Free Space:** We see 139k tokens of free space (69.6%), meaning we have plenty of capacity remaining.
* **Autocompact Buffer:** Space reserved to manage context compaction automatically as memory fills up.

As you work with Claude Code, this percentage will grow as you add files to the session, write code, and accumulate history. When usage approaches high levels (**80-90%**), Claude's responses may slow down, or you might need to use `/clear` to maintain peak performance.

## Configuration and Permissions

Two additional commands round out our essential toolkit:

### Settings Tuning via `/config`

The `/config` command allows us to view and modify our global Claude Code settings, such as setting default models, editor preferences, and session behaviors. You can also quickly access these settings by navigating to the **Config** tab within the `/status` output window.

### Guardrails via `/permissions`

The `/permissions` command manages what actions Claude Code is authorized to perform autonomously. Since Claude Code can read files, write code, and execute shell commands directly, permissions control which operations require explicit human approval.

> 🛡️ **Safety Layer:** For maximum productivity, you might allow Claude to read files freely, but enforce a prompt confirmation rule before it executes any potentially destructive terminal commands.

## Summary of Environment Controls

| Command | Primary Function | Primary Use Case |
| --- | --- | --- |
| **`/status`** | Displays cwd, session variables, and configuration sources. | Investigating path issues or locating file access bugs. |
| **`/model`** | Opens an interactive selector or switches model tiers directly. | Swapping between Haiku (low cost/boilerplate) and Sonnet (complex tasks). |
| **`/context`** | Visualizes active memory usage and remaining token budgets. | Monitoring space limits during long refactoring sessions in large projects. |
| **`/config`** | Adjusts system settings and workspace formatting parameters. | Customizing editor properties and tool behaviors. |
| **`/permissions`** | Configures automatic tool execution versus human approval rules. | Setting safety guardrails on terminal executions. |

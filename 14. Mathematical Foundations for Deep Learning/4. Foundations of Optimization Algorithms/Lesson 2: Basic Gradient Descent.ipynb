## Understanding Gradient Descent üß†  
Gradient descent is an algorithm used to minimize a function. It works similarly to Newton's method and is crucial in machine learning for optimizing loss functions that measure an algorithm's error.  

### Mathematically:  
We minimize a function \( f(x, y) \) by iteratively moving in the direction of the **negative gradient** (the steepest descent). The gradient is a vector of partial derivatives pointing toward the steepest ascent, and moving against it leads to the minimum value.  

### Analogy:
Imagine playing a game where you need to find treasure buried at the lowest point of a landscape. Gradient descent guides you by showing the slope of the land, helping you move downhill toward the treasure.

## Example: Simple Quadratic Function ‚öôÔ∏è  
Consider the function:  
\[
f(x, y) = x^2 + y^2

This bowl-shaped surface has its minimum value at the origin \((0, 0)\).  

Think of this function as a stretched rubber sheet. If you press at any point, the gradient indicates the steepest rise, and moving in the opposite direction leads to the minimum.  

### Gradient Calculation:
For \( f(x, y) = x^2 + y^2 \), the gradient is:  
\[
\nabla f(x, y) = [2x, 2y]
\]  
This tells us how the function changes at a given point.  

#### Analogy:

## Implementing Gradient Descent: Part 1 üí°  

### Formula:
\[
x_{t+1} = x_t - \eta \nabla f(x_t)
\]  

Where:  
- \( x_t \): Current point  
- \( \eta \): Learning rate  
- \( \nabla f(x_t) \): Gradient at the current point  

### Python Implementation:  
```python
# Gradient descent without path output
def simple_gradient_descent(f_grad, init_point, learning_rate=0.1, iterations=100):
    point = list(init_point)
    for _ in range(iterations):
        grad = f_grad(point)
        point = [p - learning_rate * g for p, g in zip(point, grad)]
    return point
```

### Inputs of `simple_gradient_descent`:  
1. `f_grad`: Gradient function of the target function.  
2. `init_point`: Starting point.  
3. `learning_rate`: Step size in each iteration.  
4. `iterations`: Number of updates to the point.  

---

## Choosing the Starting Point and Learning Rate ‚öôÔ∏è  
- **Starting Point**: Initial guess for the minimum. Example: \([2, 2]\).  
- **Learning Rate (\( \eta \))**: Controls step size.  
  - High \(\eta\): Risk of overshooting.  
  - Low \(\eta\): Slower convergence.

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the complex function
def complex_function(x, y):
    return x**4 - 3*x**3 + y**4 - 3*y**3

# Define the gradient of the function
def complex_gradient(x, y):
    dfdx = 4 * x**3 - 9 * x**2
    dfdy = 4 * y**3 - 9 * y**2
    return np.array([dfdx, dfdy])

# Implement the gradient descent algorithm
def gradient_descent(initial_point, learning_rate, num_iterations, tolerance=1e-6):
    point = np.array(initial_point, dtype=float)
    for i in range(num_iterations):
        gradient = complex_gradient(point[0], point[1])
        new_point = point - learning_rate * gradient

        # Check for convergence
        if np.linalg.norm(new_point - point) < tolerance:
            print(f"Converged after {i} iterations")
            return new_point

        point = new_point

    return point

# Parameters for gradient descent
initial_point = [0, 0]  # Define initial point
learning_rate = 0.01  # Define the learning rate
num_iterations = 10000

# Run gradient descent
optimal_point = gradient_descent(initial_point, learning_rate, num_iterations)
print("Optimal point after gradient descent:", optimal_point)

# Plot the contour graph of the function
x = np.linspace(-2, 3, 400)
y = np.linspace(-2, 3, 400)
X, Y = np.meshgrid(x, y)
Z = complex_function(X, Y)

plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(contour)
plt.title('Contour Plot of the Complex Function')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(optimal_point[0], optimal_point[1], color='red', label='Optimal Point')
plt.legend()
plt.show()
````

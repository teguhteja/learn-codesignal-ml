# Lesson 1: Exploring Second Derivative


Hello! Today, we'll explore the **second derivative** in calculus and its importance in machine learning. By the end of this lesson, you'll understand what the second derivative is, why it's useful, and how to compute it in Python.

In machine learning, first and second derivatives are essential for optimization. The second derivative helps us understand the curvature of a function, showing how the rate of change itself changes over time.

---

## What is the Second Derivative?

Let's start by understanding the second derivative. Recall that the **first derivative** \( f'(x) \) of a function \( f(x) \) tells us the rate at which \( f(x) \) changes. The **second derivative** \( f''(x) \) tells us how the rate of change of \( f(x) \) itself is changing:

\[
f''(x) = \frac{d}{dx}\left(\frac{df(x)}{dx}\right) = \frac{d^2f(x)}{dx^2}
\]

Think of driving a car:

- The first derivative is your **speed**.
- The second derivative is your **acceleration**.

If you're speeding up or slowing down, the second derivative shows how quickly your speed is changing.

---

## Why Do We Need the Second Derivative?

The second derivative is crucial for understanding the shape and behavior of functions. Here are two reasons why it's important:

1. **Concavity**: Indicates if the function's graph bends upward or downward.
   - When \( f''(x) > 0 \), the graph is **concave up** (like a smile).
   - When \( f''(x) < 0 \), the graph is **concave down** (like a frown).

2. **Inflection Points**: Points where the function changes concavity. This occurs when \( f''(x) = 0 \).

In machine learning, second derivatives are used in optimization techniques like **Newton's Method**, which uses the curvature of the error function to efficiently find the model parameters that minimize errors.

---

## Example Function

Let's use a polynomial function \( f(x) = x^3 - 3x^2 + 2x \) to explore derivatives.

- The **first derivative** of \( f(x) \) is:
  \[
  f'(x) = 3x^2 - 6x + 2
  \]

- The **second derivative** of \( f(x) \) is:
  \[
  f''(x) = 6x - 6
  \]

As you can see, both of them are functions of \( x \). You can calculate the third, fourth, fifth derivative, and so on, but their use-cases are quite rare.

---

## Exploring the Plot

Let's look at \( f(x) \), \( f'(x) \), and \( f''(x) \) on the same plot:

1. When the **first derivative** is zero, the original function has a **peak**: either a minimum or maximum.
2. When the **second derivative** is zero, the first derivative has a **peak** (in this case, its minimum). At this point, the original function has the **lowest rate of change**.
3. When the **second derivative** is zero, the original function changes from being **concave up** to being **concave down**.

## Plotting Sigmoid Function and Its Derivatives

The error occurs because the second derivative \( f''(x) \) is a constant value (4), which results in a single scalar value, whereas the x-values (\(x\)) are an array with 400 points. To fix this, you need to create an array of the same size as \(x\) for \(f''(x)\). Here's the corrected code:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the quadratic function f(x)
def f(x):
    return 2 * x**2 - 4 * x + 1

# Define the first derivative f'(x)
def f_prime(x):
    return 4 * x - 4

# Define the second derivative f''(x)
def f_double_prime(x):
    return np.full_like(x, 4)  # Create an array of the same shape as x, filled with the value 4

# Generate x values
x = np.linspace(-10, 10, 400)

# Calculate y values for f(x), f'(x), and f''(x)
y_f = f(x)
y_f_prime = f_prime(x)
y_f_double_prime = f_double_prime(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y_f, label="$f(x) = 2x^2 - 4x + 1$", linewidth=2)
plt.plot(x, y_f_prime, label="$f'(x) = 4x - 4$", linestyle='--', linewidth=2)
plt.plot(x, y_f_double_prime, label="$f''(x) = 4$", linestyle='-.', linewidth=2)

# Add title and labels
plt.title("Function $f(x) = 2x^2 - 4x + 1$ and Its Derivatives")
plt.xlabel("x")
plt.ylabel("y")

# Add legend
plt.legend()

# Add grid
plt.grid(True)

# Show plot
plt.show()
```

### Explanation of the Fix
- The function `f_double_prime(x)` now returns an array of the same size as `x`, filled with the constant value \(4\), using `np.full_like(x, 4)`.
- This ensures that all \(y\)-values for \(f''(x)\) match the shape of \(x\), avoiding the dimension mismatch error. 

This corrected code will plot \(f(x)\), \(f'(x)\), and \(f''(x)\) correctly without errors.

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the quadratic function f(x)
def f(x):
    return 2 * x**2 - 4 * x + 1

# Define the first derivative f'(x)
def f_prime(x):
    return 4 * x - 4

# Define the second derivative f''(x)
def f_double_prime(x):
    return np.full_like(x, 4)  # Create an array of the same shape as x, filled with the value 4

# Generate x values
x = np.linspace(-10, 10, 400)

# Calculate y values for f(x), f'(x), and f''(x)
y_f = f(x)
y_f_prime = f_prime(x)
y_f_double_prime = f_double_prime(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y_f, label="$f(x) = 2x^2 - 4x + 1$", linewidth=2)
plt.plot(x, y_f_prime, label="$f'(x) = 4x - 4$", linestyle='--', linewidth=2)
plt.plot(x, y_f_double_prime, label="$f''(x) = 4$", linestyle='-.', linewidth=2)

# Add title and labels
plt.title("Function $f(x) = 2x^2 - 4x + 1$ and Its Derivatives")
plt.xlabel("x")
plt.ylabel("y")

# Add legend
plt.legend()

# Add grid
plt.grid(True)

# Show plot
plt.show()
````

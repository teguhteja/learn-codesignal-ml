## What is a Gradient? ðŸ¤”

Imagine a landscape of hills and valleys. The **gradient** tells you the direction of the steepest climb. Formally, the gradient is a **vector** containing all the partial derivatives of a multivariable function, pointing toward the steepest increase.

### Example: Hiking Analogy
If you're climbing a hill and want to reach the top as quickly as possible, the gradient guides you on the steepest path.

In **machine learning**, gradients are crucial for:  
- **Training models**: Finding the best parameters to minimize a loss function.  
- **Optimizing weights** in neural networks for better predictions.  

Think of the gradient as a map showing the quickest way to minimize errors.

---

## Example Function and Partial Derivatives as Vectors

Consider the function:  
\[
f(x, y) = x^2 + y^2
\]  

### Partial Derivatives
For \(f(x, y) = x^2 + y^2\), the partial derivatives are:  
\[
\frac{\partial f}{\partial x} = 2x \quad \text{and} \quad \frac{\partial f}{\partial y} = 2y
\]  

### Gradient Vector
At any point \((x, y)\), the gradient is:  
\[
\nabla f(x, y) = [2x, 2y]
\]  
This vector points in the direction of the steepest increase of the function.

- The gradient vector at a point (e.g., (1, -1)) as an arrow pointing toward the steepest ascent.

The **negative gradient** points to the steepest descent, useful for minimizing the function value (e.g., optimizing machine learning models).

## Calculating the Gradient Using Central Difference Formula

### Part 1: Gradient Calculation
To approximate gradients, we use the **central difference method**, which measures the difference in values at slightly different points. Here's the Python function:

```python
# Centrally differenced gradient for multivariable functions
def gradient(f, pt, h=1e-5):
    grad = []
    for i in range(len(pt)):
        augmented_point = pt[:]
        augmented_point[i] += h
        grad.append((f(*augmented_point) - f(*pt)) / h)
    return grad
```

### Explanation
1. `grad = []`: Initializes an empty list to store partial derivatives.
2. `for i in range(len(pt))`: Iterates over variables in `pt` (e.g., \(x, y\)).
3. `augmented_point = pt[:]`: Copies `pt` to avoid modifying the original.
4. `augmented_point[i] += h`: Slightly perturbs the \(i\)-th variable.
5. `grad.append((f(*augmented_point) - f(*pt)) / h)`: Calculates and appends the partial derivative.
6. `return grad`: Returns the gradient vector.

This method works for functions with any number of variables.

---

### Part 2: Applying the Formula

Example with a multivariable function:  
\[
f(x, y) = x^2 + y^2
\]

```python
# Multivariable function example
multivariable_fn = lambda x, y: x**2 + y**2

print("Gradient of f(x, y) at (1, -1):", gradient(multivariable_fn, [1, -1]))
# Expected output: [2, -2]
```

### Output  
The computed gradient at (1, -1) is approximately \([2, -2]\), demonstrating the central difference method's accuracy.

  - \( \frac{\partial f}{\partial y} = \frac{f(x, y+h) - f(x, y)}{h} \)

- **Gradient Vector**: \([ \nabla_x, \nabla_y ] = [2.0, 2.0]\) at \( (2, 2) \).

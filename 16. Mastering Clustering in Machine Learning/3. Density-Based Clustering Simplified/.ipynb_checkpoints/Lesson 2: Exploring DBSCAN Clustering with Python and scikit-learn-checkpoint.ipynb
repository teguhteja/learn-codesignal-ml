Welcome! In this lesson, we will explore the **Density-Based Spatial Clustering of Applications with Noise (DBSCAN)** algorithm. Using Python's `sklearn.cluster` library, we will implement DBSCAN and use **Matplotlib** to visualize its output. Ready to dive into DBSCAN? Let’s get started!

# **Essential Python Libraries**

To begin, we need to import the necessary Python libraries:

- **`numpy`**: For matrix computation.
- **`sklearn.cluster`**: For DBSCAN implementation.
- **`sklearn.datasets`**: For generating synthetic data.
- **`Matplotlib`**: For visualizing the clusters.

```python
from sklearn.cluster import DBSCAN
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
```

---

# **Creating a Synthetic Dataset**

Next, we use the `make_blobs` function from `sklearn.datasets` to generate synthetic clusters. This function creates isotropic Gaussian blobs for clustering.

```python
# Configuration options
num_samples_total = 180
cluster_centers = [(3, 3), (7, 7)]
num_classes = len(cluster_centers)

# Generate clusters
data, y = make_blobs(
    n_samples=num_samples_total,
    centers=cluster_centers,
    n_features=num_classes,
    center_box=(0, 1),
    cluster_std=0.5,
    random_state=42
)
```

In the code above:
- We generate **180 samples** distributed among clusters centered at `(3, 3)` and `(7, 7)`.
- The parameter `center_box=(0, 1)` ensures that the cluster centers are within the square `[0, 0]` to `[1, 1]`.

---

# **Running DBSCAN**

With the data ready, we can now run the **DBSCAN algorithm**. DBSCAN requires two key parameters:
- **`eps`**: The maximum distance between two samples to be considered in the same neighborhood.
- **`min_samples`**: The minimum number of samples in a neighborhood for a point to be considered a core point.

```python
# Parameters
epsilon = 0.5
min_samples = 13

# Run DBSCAN
dbscan = DBSCAN(eps=epsilon, min_samples=min_samples)
dbscan.fit(data)
```

In this example:
- We set `eps` to **0.5** and `min_samples` to **13**.
- The DBSCAN object is initialized and fitted to the dataset.

To count the number of clusters formed (excluding noise points, which are labeled as `-1` by DBSCAN), we use the following code:

```python
# Number of clusters in labels, ignoring noise if present
n_clusters_ = len(set(dbscan.labels_)) - (1 if -1 in dbscan.labels_ else 0)
print('Estimated number of clusters: %d' % n_clusters_)  # Output: 2
```

---

# **Visualizing DBSCAN Clusters with Matplotlib**

Finally, we visualize the output of the DBSCAN clustering using **Matplotlib**. Each cluster is color-coded, and noise points are displayed in black.

```python
# Identify outliers (noise points)
outliers = data[dbscan.labels_ == -1]

# Generate colors for clusters
colors = plt.cm.rainbow(np.linspace(0, 1, len(np.unique(dbscan.labels_))))

# Plot clusters
for i in range(len(data)):
    plt.plot(data[i][0], data[i][1], 'o', color=colors[int(dbscan.labels_[i] % len(colors))])

# Plot noise points
plt.plot(outliers[:, 0], outliers[:, 1], 'o', color='black')

plt.show()
```

### Explanation:
- Each point represents a data sample, and its color corresponds to the cluster it belongs to.
- The **rainbow colormap** is used to assign unique colors to each cluster.
- Noise points (outliers) are displayed in **black**.

### **Output**
When you run the code:

- Noise points (if any) will be shown in black.
- The title will indicate the estimated number of clusters.

This example highlights the power of DBSCAN in clustering non-linear and complex datasets, making it a great choice for tasks involving irregularly shaped clusters.

````python
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import numpy as np

# Generate a moon-shaped dataset and standardize features
X, y = make_moons(n_samples=200, noise=0.05, random_state=42)
X = StandardScaler().fit_transform(X)

# Implement DBSCAN
db = DBSCAN(eps=0.3, min_samples=5).fit(X)

# Extract labels
labels = db.labels_

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

# Plotting the clusters
colors = plt.cm.rainbow(np.linspace(0, 1, len(set(labels))))

for i in range(len(X)):
    if labels[i] == -1:
        plt.plot(X[i][0], X[i][1], 'o', markerfacecolor='k', markeredgecolor='k')
    else:
        plt.plot(X[i][0], X[i][1], 'o', markerfacecolor=colors[labels[i]], markeredgecolor=colors[labels[i]])

plt.title('Estimated number of clusters: %d' % n_clusters_)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.show()
````

     - `min_samples=5`: Minimum number of points required to form a dense region.

3. **Fitting the Model**:
   - `db.fit(X)` fits the DBSCAN model to the dataset.
   - `labels = db.labels_` retrieves the cluster labels for each data point. Noise points are labeled as `-1`.

4. **Core Samples**:
   - `core_samples_mask` identifies which points are core samples (part of dense regions).

5. **Visualization**:
   - Each cluster is assigned a unique color using `plt.cm.Spectral`.
   - Core points are plotted larger (`markersize=14`), while non-core points are smaller (`markersize=6`).
   - Noise points are plotted in black.

### What to Observe:
- **Clusters**: Points in the same cluster will have the same color.
- **Noise Points**: Points classified as noise will appear in black.
- **Core vs. Non-Core Points**: Core points are larger, while non-core points are smaller.

Run the code and enjoy visualizing the density-based clusters in your dataset! 🚀

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

# Create a toy dataset with two interleaved half circles
X, _ = make_moons(n_samples=100, noise=0.05)  # Reduced the number of samples for simplicity

# Initialize the DBSCAN clustering model with parameters eps=0.3 and min_samples=5
db = DBSCAN(eps=0.3, min_samples=5)

# Apply the model to the dataset and get the cluster labels for each data point
db.fit(X)
labels = db.labels_

# Identify which data points are core samples
core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True

# Generating colors for the clusters
unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

# Plotting
for k, col in zip(unique_labels, colors):
    if k == -1:  # Noise points are labeled -1
        col = [0, 0, 0, 1]  # Black color for noise

    class_member_mask = (labels == k)
    xy = X[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=14)

    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=6)

plt.title('DBSCAN: Visualizing Density-Based Clusters')
plt.show()
````

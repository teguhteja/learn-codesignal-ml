# Lesson 3: Tuning DBSCAN Parameters

Explore how parameter changes in the DBSCAN algorithm can alter pattern recognition in astronomical data, revealing different constellations in the cosmic landscape. The provided code visualizes clustering results on a moon-shaped dataset using various DBSCAN parameter sets. Click Run to observe how each set of parameters reveals a unique clustering configuration!

```py
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

# Generate moon-like dataset
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)

# Run DBSCAN with different eps, min_samples, metric, and algorithm
db1 = DBSCAN(eps=0.3, min_samples=5, metric='euclidean', algorithm='auto').fit(X)
db2 = DBSCAN(eps=0.3, min_samples=10, metric='manhattan', algorithm='ball_tree').fit(X)
db3 = DBSCAN(eps=0.4, min_samples=5, metric='cosine', algorithm='brute').fit(X)

# Visualize the results
plt.figure(figsize=(20, 5))

plt.subplot(131)
plt.scatter(X[:,0], X[:,1], c=db1.labels_)
plt.title('eps=0.3, min_samples=5, euclidean, auto')

plt.subplot(132)
plt.scatter(X[:,0], X[:,1], c=db2.labels_)
plt.title('eps=0.3, min_samples=10, manhattan, ball_tree')

plt.subplot(133)
plt.scatter(X[:,0], X[:,1], c=db3.labels_)
plt.title('eps=0.4, min_samples=5, cosine, brute')

plt.tight_layout()
plt.show()

```
Below is an overview of how each parameter in DBSCAN contributes to the resulting cluster patterns, especially in a context such as astronomical data analysis. In your example, we have three DBSCAN configurations—each produces unique clustering results thanks to variations in the DBSCAN hyperparameters: eps, min_samples, metric, and algorithm.

1. eps (ε): The Neighborhood Radius
   • This parameter represents the maximum distance between two samples for them to be considered neighbors.
   • A smaller eps (e.g., eps=0.3) restricts neighbor searches to nearby points, potentially fragmenting data into multiple smaller clusters or leaving more points labeled as noise (-1).
   • A larger eps (e.g., eps=0.4) broadens neighbor searches, which can merge adjacent clusters, or reduce the number of points classified as noise.

2. min_samples: Minimum Cluster Size
   • Denotes the minimum number of points required to form a dense region.
   • When min_samples is low (e.g., min_samples=5), clusters can emerge around moderately dense regions.
   • When min_samples is high (e.g., min_samples=10), DBSCAN becomes stricter, so it may merge points into fewer clusters or classify more points as noise if the necessary density is not met.

3. metric: Distance Function
   • Determines how distance between points is measured.
   • Euclidean: Straight-line (L2-norm) distance.
   • Manhattan: Sum of absolute coordinate differences (L1-norm).
   • Cosine: Computes similarity based on the angle between vectors, resulting in different cluster boundaries when points differ primarily in direction rather than location.
   • Changing the metric can drastically alter cluster shapes. For instance, if your data has strong directional components or outliers, switching from Euclidean to Cosine distance might highlight patches of data aligned in a particular “direction” in the feature space.

4. algorithm: Nearest Neighbors Search Strategy
   • 'auto': Lets scikit-learn decide the most efficient method based on data, metric, and other factors.
   • 'ball_tree': Uses a spatial data structure specialized for certain distance metrics (e.g., Euclidean, Manhattan).
   • 'kd_tree': Similar to ball_tree but structured differently.
   • 'brute': Performs a brute-force distance calculation, appropriate for smaller datasets or certain distance metrics (e.g., cosine, which may not be compatible with kd_tree).

Putting It All Together in Your Example
• db1 (eps=0.3, min_samples=5, metric='euclidean', algorithm='auto'):
  - A moderately small radius and low minimum cluster size.
  - Likely to produce well-defined clusters with outliers or small noises.
  - Euclidean distance and the default algorithm (auto) are a common starting point.

• db2 (eps=0.3, min_samples=10, metric='manhattan', algorithm='ball_tree'):
  - Same radius as db1 but higher minimum cluster size, thus requiring more samples in a neighborhood before forming a cluster.
  - Manhattan distance can separate clusters differently, particularly if the data’s distribution suggests “grid-like” separations rather than diagonal/straight lines.
  - ball_tree is suitable for metrics like Manhattan on medium-to-large datasets.

• db3 (eps=0.4, min_samples=5, metric='cosine', algorithm='brute'):
  - A larger neighborhood radius might “merge” clusters more than the smaller radius in db1/db2.
  - The same low min_samples (5) can keep clusters more flexible.
  - Cosine distance measures angular difference, which can generate unexpectedly shaped clusters if points align directionally.
  - brute is chosen because cosine often lacks efficient tree-based neighbor search support.

Why This Matters for Astronomical Data (“Constellations” in the Cosmic Landscape)
• Astronomical data is often high-dimensional and can exhibit complex features—light intensity, spectrum lines, or positions in multi-dimensional space.
• A Euclidean distance might group together objects that are close by straightforward geometric proximity.
• A Cosine distance can reveal objects that share a similar “direction” in feature space (e.g., similar spectral profiles but different magnitudes).
• min_samples and eps help define how tightly or loosely distributed celestial bodies should be before they form a “constellation” (cluster).
• Depending on your goals—finding distinct galaxies, star clusters, or other cosmic structures—you might experiment with these parameters (including other distance metrics) to refine which astronomical objects belong together.

In short, DBSCAN’s flexibility allows for different ways to reveal and interpret structure in data that might otherwise appear as noise or random scatter from a single perspective. By adjusting eps, min_samples, metric, and algorithm, you can tailor clustering to highlight different “constellations” in your own cosmic or moon-shaped datasets.

• In practice, if your data has a “grid-like” spread or if features differ in magnitude, Manhattan distance may highlight those structures more distinctly.  
• Adjusting only one parameter at a time (like switching from 'cosine' to 'manhattan') is a great way to see how each setting influences clustering. As you’ll observe, even small changes can yield notable shifts in how DBSCAN forms groups—or identifies outliers—in your dataset.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

# Generate moon-like dataset
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)

# Run DBSCAN with different parameters
db1 = DBSCAN(eps=0.3, min_samples=5, metric='euclidean', algorithm='auto').fit(X)
db2 = DBSCAN(eps=0.3, min_samples=10, metric='manhattan', algorithm='ball_tree').fit(X)
db3 = DBSCAN(eps=0.4, min_samples=5, metric='cosine', algorithm='brute').fit(X)

# Visualize the results
plt.figure(figsize=(15, 5))
plt.subplot(131)
plt.scatter(X[:,0], X[:,1], c=db1.labels_)
plt.title('DBSCAN: eps=0.3, min_samples=5, euclidean, auto')

plt.subplot(132)
plt.scatter(X[:,0], X[:,1], c=db2.labels_)
plt.title('DBSCAN: eps=0.5, min_samples=10, manhattan, ball_tree')  # Learn to change 'manhattan' to 'euclidean'

plt.subplot(133)
plt.scatter(X[:,0], X[:,1], c=db3.labels_)
plt.title('DBSCAN: eps=0.4, min_samples=5, cosine, brute')

plt.tight_layout()
plt.show()
````

Why These Choices?
--------------------------------------------------------------------------------

• db2 uses a small eps (0.1) and a higher min_samples (15), making the clustering criteria stricter. Combined with the cosine metric, only points that are close in angular space will form clusters.  
• db3 uses a larger eps (0.5), which can merge points more readily into bigger clusters, while the manhattan metric (also called L1 distance) measures “grid-like” distance, and ball_tree efficiently handles queries for many distance calculations in medium to large data.  

By experimenting with different distance metrics, nearest neighbor algorithms, and hyperparameter settings for eps and min_samples, you can uncover nuances in how DBSCAN discovers or merges clusters—much like revealing distinct “constellations” in your dataset!

```py
from sklearn.datasets import make_moons

# Generate sample data
X, _ = make_moons(n_samples=200, noise=0.1, random_state=42)

# 1) DBSCAN with Manhattan distance, eps=0.2, min_samples=5
dbscan_manhattan = DBSCAN(eps=0.2, min_samples=5, metric='manhattan')
dbscan_manhattan.fit(X)

# 2) DBSCAN with Euclidean distance, eps=0.2, min_samples=5
dbscan_euclidean = DBSCAN(eps=0.2, min_samples=5, metric='euclidean')
dbscan_euclidean.fit(X)

# 3) DBSCAN with Cosine distance, eps=0.1, min_samples=55
dbscan_cosine = DBSCAN(eps=0.1, min_samples=55, metric='cosine')
dbscan_cosine.fit(X)

# Compare the clustering results
plt.figure(figsize=(15, 5))

# Plot Manhattan
plt.subplot(131)
plt.scatter(X[:, 0], X[:, 1], c=dbscan_manhattan.labels_)
plt.title('DBSCAN with Manhattan distance\n(eps=0.2, min_samples=5)')

# Plot Euclidean
plt.subplot(132)
plt.scatter(X[:, 0], X[:, 1], c=dbscan_euclidean.labels_)
plt.title('DBSCAN with Euclidean distance\n(eps=0.2, min_samples=5)')

# Plot Cosine
plt.subplot(133)
plt.scatter(X[:, 0], X[:, 1], c=dbscan_cosine.labels_)
plt.title('DBSCAN with Cosine distance\n(eps=0.1, min_samples=55)')

plt.tight_layout()
plt.show()
```

--------------------------------------------------------------------------------
How It Works
--------------------------------------------------------------------------------
• The first DBSCAN (dbscan_manhattan) uses “manhattan” distance, which considers the sum of absolute differences along each axis.
• The second DBSCAN (dbscan_euclidean) switches to the common Euclidean distance, measuring straight-line distances in the 2D plane.
• The third DBSCAN (dbscan_cosine) uses the “cosine” distance, focusing on angular differences between points, which often highlights a different notion of similarity.

By plotting all three results side-by-side, you can see the impact of each metric—and the chosen eps and min_samples—on how DBSCAN identifies clusters and noise in the data.

````python
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

# Generate sample data
X, _ = make_moons(n_samples=200, noise=0.1, random_state=42)

# 1) DBSCAN with Manhattan distance, eps=0.2, min_samples=5
dbscan_manhattan = DBSCAN(eps=0.2, min_samples=5, metric='manhattan')
dbscan_manhattan.fit(X)

# 2) DBSCAN with Euclidean distance, eps=0.2, min_samples=5
dbscan_euclidean = DBSCAN(eps=0.2, min_samples=5, metric='euclidean')
dbscan_euclidean.fit(X)

# 3) DBSCAN with Cosine distance, eps=0.1, min_samples=55
dbscan_cosine = DBSCAN(eps=0.1, min_samples=55, metric='cosine')
dbscan_cosine.fit(X)

# Compare the clustering results
plt.figure(figsize=(15, 5))

# Plot Manhattan
plt.subplot(131)
plt.scatter(X[:, 0], X[:, 1], c=dbscan_manhattan.labels_)
plt.title('DBSCAN with Manhattan distance\n(eps=0.2, min_samples=5)')

# Plot Euclidean
plt.subplot(132)
plt.scatter(X[:, 0], X[:, 1], c=dbscan_euclidean.labels_)
plt.title('DBSCAN with Euclidean distance\n(eps=0.2, min_samples=5)')

# Plot Cosine
plt.subplot(133)
plt.scatter(X[:, 0], X[:, 1], c=dbscan_cosine.labels_)
plt.title('DBSCAN with Cosine distance\n(eps=0.1, min_samples=55)')

plt.tight_layout()
plt.show()
````

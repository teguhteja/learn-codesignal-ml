# Dendrograms and Hierarchical Clustering

## Introduction
Welcome! Following our exploration of Hierarchical Clustering, we'll examine dendrograms - visual tools that illuminate hierarchical clustering in an aesthetically pleasing form. We'll learn to read, analyze, and interpret dendrograms using Python.

## Scipy's linkage() for Hierarchical Clustering
The Scipy library's `linkage()` function performs agglomerative hierarchical clustering, taking an array of data points and the clustering method as primary inputs. Here's how to perform Hierarchical Agglomerative clustering on our cities dataset:

```python
import numpy as np
from scipy.cluster.hierarchy import linkage

# Data provided
data = {
    'city': ['Mexico City', 'Los Angeles', 'Bogota', 'Lima', 'Madrid', 
             'Berlin', 'Kinshasa', 'Lagos', 'New Delhi', 'Beijing', 'Sydney'],
    'long': [-99.1332, -118.2437, -74.0721, -77.0428, -3.7038, 
             13.4050, 15.2663, 3.3792, 77.2090, 116.4074, 151.2093],
    'lat': [19.4326, 34.0522, 4.7110, -12.0464, 40.4168, 
            52.5200, -4.4419, 6.5244, 28.6139, 39.9042, -33.8688]
}

# Convert data to an array of coordinates
coordinates = np.column_stack((data['long'], data['lat']))

# Perform hierarchical clustering
Z = linkage(coordinates, method='ward')
```

## Plotting Dendrograms
```python
from scipy.cluster.hierarchy import dendrogram
import matplotlib.pyplot as plt

# Plotting the dendrogram
plt.figure(figsize=(10, 8))
dendrogram(Z, labels=data['city'], leaf_rotation=90, leaf_font_size=12)
plt.title('Dendrogram of Cities based on Geographic Coordinates')
plt.xlabel('Cities')
plt.ylabel('Distance')
plt.show()
```

## Interpreting Dendrograms
- The x-axis shows individual data points (city names)
- The y-axis represents distance/dissimilarity between clusters
- Higher horizontal lines indicate larger distances between clusters
- Moving bottom to top shows progression from individual points to complete clusters
- Cutting the dendrogram at specific heights yields different numbers of clusters

## Key Differences: linkage() vs AgglomerativeClustering
- `linkage()`: Better for smaller datasets, provides quick access to full dendrogram
- `AgglomerativeClustering`: Suited for large datasets, memory efficient, flexible cluster specification

```python
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'city': ['Kyoto', 'Oslo', 'Cape Town', 'Rio', 'Moscow', 'Paris', 'Dubai', 'Toronto', 'Seoul', 'Shanghai'],
    'long': [135.7681, 10.7522, 18.4241, -43.1729, 37.6173, 2.3522, 55.2708, -79.3832, 126.9780, 121.4737],
    'lat': [35.0116, 59.9139, -33.9249, -22.9068, 55.7558, 48.8566, 25.2048, 43.6532, 37.5665, 31.2304]
}

# Convert data to an array of coordinates
coordinates = np.column_stack((data['long'], data['lat']))

# Perform hierarchical clustering with 'complete' linkage
Z = linkage(coordinates, method='complete')  # Changed from 'ward' to 'complete'

# Plotting the dendrogram
plt.figure(figsize=(10, 8))
dendrogram(Z, labels=data['city'], leaf_rotation=90, leaf_font_size=12)
plt.title('Dendrogram of Cities using Complete Linkage Method')
plt.xlabel('Cities')
plt.ylabel('Distance')
plt.show()
```

Key differences when using complete linkage:

1. Complete linkage defines cluster distance as the maximum distance between any two points in different clusters, which can lead to more spread-out clusters

2. We might observe:
   - Different city groupings compared to Ward's method
   - Potentially longer vertical lines, indicating larger distances between clusters
   - More distinct separation between some clusters

The complete linkage method tends to create more evenly sized clusters and is less sensitive to noise, though it might sometimes struggle with elongated clusters. Would you like me to explain any specific patterns you notice in the resulting dendrogram?

````python
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'city': ['Kyoto', 'Oslo', 'Cape Town', 'Rio', 'Moscow', 'Paris', 'Dubai', 'Toronto', 'Seoul', 'Shanghai'],
    'long': [135.7681, 10.7522, 18.4241, -43.1729, 37.6173, 2.3522, 55.2708, -79.3832, 126.9780, 121.4737],
    'lat': [35.0116, 59.9139, -33.9249, -22.9068, 55.7558, 48.8566, 25.2048, 43.6532, 37.5665, 31.2304]
}

# Convert data to an array of coordinates
coordinates = np.column_stack((data['long'], data['lat']))

# Perform hierarchical clustering with 'complete' linkage
Z = linkage(coordinates, method='complete')  # Changed from 'ward' to 'complete'

# Plotting the dendrogram
plt.figure(figsize=(10, 8))
dendrogram(Z, labels=data['city'], leaf_rotation=90, leaf_font_size=12)
plt.title('Dendrogram of Cities using Complete Linkage Method')
plt.xlabel('Cities')
plt.ylabel('Distance')
plt.show()
````

```python
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Example dataset of 2D points.
data = {
    "point": ["A", "B", "C", "D"],
    "x": [1.0, 2.0, 3.0, 4.0],
    "y": [4.0, 3.5, 2.0, 1.5]
}

# Convert data to an array of coordinates.
coordinates = np.column_stack((data["x"], data["y"]))

# Perform hierarchical clustering using the 'ward' method.
Z = linkage(coordinates, method='ward')

# Create the dendrogram visualization
plt.figure(figsize=(8, 6))
dendrogram(Z, 
    labels=data["point"],          # Use point labels A, B, C, D
    leaf_rotation=90,              # Rotate labels for better readability
    leaf_font_size=12              # Set font size for clarity
)

# Add title and labels
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Points')
plt.ylabel('Distance')

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Display the plot
plt.show()
```

This visualization will show:
- Clear labels (A, B, C, D) on the x-axis
- Distance measurements on the y-axis
- The hierarchical relationships between points
- A clean, readable layout with rotated labels

The dendrogram will reveal how the points cluster together based on their Euclidean distances, with the height of each connection indicating the distance between merged clusters.

````python
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Example dataset of 2D points.
data = {
    "point": ["A", "B", "C", "D"],
    "x": [1.0, 2.0, 3.0, 4.0],
    "y": [4.0, 3.5, 2.0, 1.5]
}

# Convert data to an array of coordinates.
coordinates = np.column_stack((data["x"], data["y"]))

# Perform hierarchical clustering using the 'ward' method.
Z = linkage(coordinates, method='ward')

# Create the dendrogram visualization
plt.figure(figsize=(8, 6))
dendrogram(Z, 
    labels=data["point"],          # Use point labels A, B, C, D
    leaf_rotation=90,              # Rotate labels for better readability
    leaf_font_size=12              # Set font size for clarity
)

# Add title and labels
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Points')
plt.ylabel('Distance')

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Display the plot
plt.show()
````

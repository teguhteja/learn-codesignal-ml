## Reading Data: Fetching a Single Document

Retrieving a document from Firestore is straightforward when you know the document's ID. For example, suppose you have a `movies` collection and want to fetch the document with the ID `the-big-new-movie-2016`.

```python
from google.cloud import firestore

db = firestore.Client()

# Reference to the document
doc_ref = db.collection('movies').document('the-big-new-movie-2016')

# Fetch the document
doc = doc_ref.get()

if doc.exists:
    print("Movie found:", doc.to_dict())
else:
    print("Movie not found.")

```

**Example Output:**

> `Movie found: {'title': 'The Big New Movie', 'genre': 'Action', 'year': 2016}`

### Selecting Specific Fields

If you only need certain fields, you can use the `field_paths` parameter. This minimizes data transfer and improves performance.

```python
# Fetch only specific fields
doc = doc_ref.get(field_paths=['title', 'genre'])

if doc.exists:
    print("Selected fields of the movie:", doc.to_dict())

```

**Example Output:**

> `Selected fields of the movie: {'title': 'The Big New Movie', 'genre': 'Action'}`

## Retrieving Multiple Documents by IDs

Sometimes you need to fetch several documents at once. Firestore provides the `get_all` method on the client object to handle this efficiently in a single request.

```python
# List of document IDs to fetch
doc_ids = [
    'the-big-new-movie-2016',
    'the-bigger-newer-movie-2017'
]

# Create document references
doc_refs = [db.collection('movies').document(doc_id) for doc_id in doc_ids]

# Use the client's get_all method to fetch all documents in one request
docs = db.get_all(doc_refs)

for doc in docs:
    if doc.exists:
        print("Movie found:", doc.id, doc.to_dict())
    else:
        print("Movie not found:", doc.id)

```

**Example Output:**

> `Movie found: the-big-new-movie-2016 {'title': 'The Big New Movie', 'genre': 'Action', 'year': 2016}`
> `Movie found: the-bigger-newer-movie-2017 {'title': 'The Bigger Newer Movie', 'genre': 'Comedy', 'year': 2017}`

> **Note:** `get_all` is a method on the **Firestore client (`db`)**, not on a collection. This allows you to retrieve multiple documents even if they belong to different collections, provided you have their references.

```python
from google.cloud import firestore

# Create Firestore client
db = firestore.Client()

# Reference to the Movies collection
movies_collection = db.collection('Movies')

# Insert movies into the collection
movies_collection.document('2016_The Big New Movie').set({
    'year': 2016,
    'title': 'The Big New Movie',
    'genre': 'Action'
})
movies_collection.document('2017_The Bigger, Newer Movie').set({
    'year': 2017,
    'title': 'The Bigger, Newer Movie',
    'genre': 'Comedy'
})
movies_collection.document('2017_Yet Another Movie').set({
    'year': 2017,
    'title': 'Yet Another Movie',
    'genre': 'Drama'
})
movies_collection.document('2017_One More Movie').set({
    'year': 2017,
    'title': 'One More Movie',
    'genre': 'Thriller'
})
movies_collection.document('2015_An Old Movie').set({
    'year': 2015,
    'title': 'An Old Movie',
    'genre': 'Classic'
})

# Get a document using its document ID
doc_ref = movies_collection.document('2016_The Big New Movie')
doc = doc_ref.get()
if doc.exists:
    print(doc.to_dict())

# Get another document with specific fields using field projection
doc_ref = movies_collection.document('2017_The Bigger, Newer Movie')
doc = doc_ref.get(field_paths=['title'])
if doc.exists:
    print(doc.to_dict())

# --- TODO FIXED: Add get_all operation to fetch both movies simultaneously ---
print("\n--- Executing Batch get_all Operation ---")

# 1. Construct the list of document references
targets = [
    movies_collection.document('2016_The Big New Movie'),
    movies_collection.document('2017_The Bigger, Newer Movie')
]

# 2. Pass references to the client's get_all method
batch_snapshots = db.get_all(targets)

# 3. Iterate over the snapshots and print their contents
for snapshot in batch_snapshots:
    if snapshot.exists:
        print(f"Retrieved via get_all: {snapshot.id} => {snapshot.to_dict()}")
    else:
        print(f"Document {snapshot.id} does not exist.")

```

---

### Mechanics of `db.get_all()`

* **Client-Level Scope:** Notice that `get_all` is called on `db` (the `firestore.Client` object), not on `movies_collection`. Because it works at the client level, you could theoretically pass document references from completely different collections (e.g., a movie document reference and a user document reference) into the same list, and Firestore would fetch them all together.
* **Order Preservation:** The snapshots returned by `get_all()` are guaranteed to match the order of the references you passed into the list.
* **Network Optimization:** Instead of triggering multiple independent HTTP request-response cycles, the SDK pipes the request as a single transactional payload, minimizing cloud infrastructure overhead and application latency.

```python
from google.cloud import firestore

# Create Firestore client
db = firestore.Client()

# Reference to the Movies collection
movies_collection = db.collection('Movies')

# Insert movies into the collection
movies_collection.document('2016_The Big New Movie').set({
    'year': 2016,
    'title': 'The Big New Movie',
    'genre': 'Action'
})
movies_collection.document('2017_The Bigger, Newer Movie').set({
    'year': 2017,
    'title': 'The Bigger, Newer Movie',
    'genre': 'Comedy'
})
movies_collection.document('2017_Yet Another Movie').set({
    'year': 2017,
    'title': 'Yet Another Movie',
    'genre': 'Drama'
})
movies_collection.document('2017_One More Movie').set({
    'year': 2017,
    'title': 'One More Movie',
    'genre': 'Thriller'
})
movies_collection.document('2015_An Old Movie').set({
    'year': 2015,
    'title': 'An Old Movie',
    'genre': 'Classic'
})
movies_collection.document('2018_Another New Movie').set({
    'year': 2018,
    'title': 'Another New Movie',
    'genre': 'Sci-Fi'
})

# --- TODO FIXED: Implement Retrieval Operations ---

print("--- 1. Simple Document Get ---")
# Retrieve the full 2016 movie document
doc_ref_2016 = movies_collection.document('2016_The Big New Movie')
doc_simple = doc_ref_2016.get()
if doc_simple.exists:
    print(f"Simple Fetch Data: {doc_simple.to_dict()}")


print("\n--- 2. Get with Field Projection ---")
# Pass a list of targeted keys to field_paths to avoid downloading unnecessary fields
doc_projected = doc_ref_2016.get(field_paths=['title', 'genre'])
if doc_projected.exists:
    print(f"Projected Fetch Data: {doc_projected.to_dict()}")


print("\n--- 3. Strongly Consistent Point Read ---")
# Direct document reads by ID are inherently strongly consistent in Firestore. 
# No special options are needed to guarantee up-to-date values.
doc_consistent = doc_ref_2016.get()
if doc_consistent.exists:
    print(f"Strongly Consistent View: {doc_consistent.to_dict()}")


print("\n--- 4. Batch get_all Consistent Read ---")
# Define the collection elements we want to batch process
doc_ref_2017 = movies_collection.document('2017_The Bigger, Newer Movie')
doc_refs = [doc_ref_2016, doc_ref_2017]

# db.get_all fetches multiple documents simultaneously, preserving strong consistency
docs_batch = db.get_all(doc_refs)
for doc in docs_batch:
    if doc.exists:
        print(f"Batch Consistent Record: {doc.id} => {doc.to_dict()}")

```

* **Batch Operations (`db.get_all`):** Passing an array of references to the base client forces the SDK to combine the requests into a single network stream. This avoids the high transactional latency penalty of nesting single `.get()` calls inside loops.
* **Guaranteed Consistency:** Because Firestore relies on consensus-based replication layers, whenever your code queries explicitly by document IDs, you are guaranteed to completely bypass stale node synchronization delays.

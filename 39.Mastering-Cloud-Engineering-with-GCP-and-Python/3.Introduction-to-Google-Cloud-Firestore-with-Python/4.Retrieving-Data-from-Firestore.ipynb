## Reading Data: Fetching a Single Document

Retrieving a document from Firestore is straightforward when you know the document's ID. For example, suppose you have a `movies` collection and want to fetch the document with the ID `the-big-new-movie-2016`.

```python
from google.cloud import firestore

db = firestore.Client()

# Reference to the document
doc_ref = db.collection('movies').document('the-big-new-movie-2016')

# Fetch the document
doc = doc_ref.get()

if doc.exists:
    print("Movie found:", doc.to_dict())
else:
    print("Movie not found.")

```

**Example Output:**

> `Movie found: {'title': 'The Big New Movie', 'genre': 'Action', 'year': 2016}`

### Selecting Specific Fields

If you only need certain fields, you can use the `field_paths` parameter. This minimizes data transfer and improves performance.

```python
# Fetch only specific fields
doc = doc_ref.get(field_paths=['title', 'genre'])

if doc.exists:
    print("Selected fields of the movie:", doc.to_dict())

```

**Example Output:**

> `Selected fields of the movie: {'title': 'The Big New Movie', 'genre': 'Action'}`

## Retrieving Multiple Documents by IDs

Sometimes you need to fetch several documents at once. Firestore provides the `get_all` method on the client object to handle this efficiently in a single request.

```python
# List of document IDs to fetch
doc_ids = [
    'the-big-new-movie-2016',
    'the-bigger-newer-movie-2017'
]

# Create document references
doc_refs = [db.collection('movies').document(doc_id) for doc_id in doc_ids]

# Use the client's get_all method to fetch all documents in one request
docs = db.get_all(doc_refs)

for doc in docs:
    if doc.exists:
        print("Movie found:", doc.id, doc.to_dict())
    else:
        print("Movie not found:", doc.id)

```

**Example Output:**

> `Movie found: the-big-new-movie-2016 {'title': 'The Big New Movie', 'genre': 'Action', 'year': 2016}`
> `Movie found: the-bigger-newer-movie-2017 {'title': 'The Bigger Newer Movie', 'genre': 'Comedy', 'year': 2017}`

> **Note:** `get_all` is a method on the **Firestore client (`db`)**, not on a collection. This allows you to retrieve multiple documents even if they belong to different collections, provided you have their references.

```python
from google.cloud import firestore

# Create Firestore client
db = firestore.Client()

# Reference to the Movies collection
movies_collection = db.collection('Movies')

# Insert movies into the collection
movies_collection.document('2016_The Big New Movie').set({
    'year': 2016,
    'title': 'The Big New Movie'
})
movies_collection.document('2017_The Bigger, Newer Movie').set({
    'year': 2017,
    'title': 'The Bigger, Newer Movie'
})
movies_collection.document('2017_Yet Another Movie').set({
    'year': 2017,
    'title': 'Yet Another Movie'
})
movies_collection.document('2017_One More Movie').set({
    'year': 2017,
    'title': 'One More Movie'
})
movies_collection.document('2015_An Old Movie').set({
    'year': 2015,
    'title': 'An Old Movie'
})
movies_collection.document('2018_Another New Movie').set({
    'year': 2018,
    'title': 'Another New Movie'
})

# Get a single document
doc_ref = movies_collection.document('2016_The Big New Movie')
doc = doc_ref.get()
if doc.exists:
    print(doc.to_dict())

# Get multiple documents using batch get
doc_refs = [
    movies_collection.document('2016_The Big New Movie'),
    movies_collection.document('2017_The Bigger, Newer Movie')
]
docs = db.get_all(doc_refs)
result_list = []
for doc in docs:
    if doc.exists:
        result_list.append(doc.to_dict())
print(result_list)

```

This script demonstrates the full lifecycle of data retrieval in Firestore—from precise document targeting to high-efficiency batch fetching. By analyzing the output, you can see how Firestore transforms document references into Python dictionaries.

### Workflow Breakdown

1. **Document ID Strategy**: Notice the naming convention used: `YYYY_Title`. This is a common pattern in Firestore to ensure document IDs are both unique and provide a hint about the data they contain without even opening the document.
2. **The `.get()` Method**: This is your surgical tool. It retrieves exactly one document. The `doc.exists` check is a "safety first" practice; in a real application, a document might be deleted between the time you create the reference and the time you call `.get()`.
3. **`db.get_all()`**: This is the performance powerhouse.
* Instead of calling `.get()` inside a loop (which creates multiple network round-trips), you pass a list of references to the client.
* The database processes these in parallel and returns them in a single stream.
* **Crucial Note**: `get_all()` returns an iterator. To see the data, you must loop through it and call `.to_dict()` on each item.

### Understanding the Output

When you run the script, the first print statement will show a single dictionary representing "The Big New Movie". The second print will show a **List of Dictionaries**.

| Method | Result Type | Best For... |
| --- | --- | --- |
| `.get()` | `DocumentSnapshot` | Specific lookups (e.g., fetching a user profile by ID). |
| `db.get_all()` | `Generator` | Multi-item lookups (e.g., fetching all items in a shopping cart). |

### Key Concept: Snapshots

When you retrieve data, you are getting a **Snapshot**. This is a read-only "frozen" image of the data at the exact moment the query was executed. If someone else updates the "Movies" collection a millisecond later, your snapshot will not change unless you perform a fresh retrieval.

To master Firestore retrieval, you need to differentiate between fetching a full document and fetching a **projection** (specific fields). Using `field_paths` is a best practice for production apps as it reduces the payload size and speeds up response times.

Here is the completed script implementing the single fetch, field selection, and batch retrieval:

```python
from google.cloud import firestore

# Create Firestore client
db = firestore.Client()

# Reference to the Movies collection
movies_collection = db.collection('Movies')

# Insert movies into the collection
movies_collection.document('2016_The Big New Movie').set({
    'year': 2016,
    'title': 'The Big New Movie',
    'genre': 'Action'
})
movies_collection.document('2017_The Bigger, Newer Movie').set({
    'year': 2017,
    'title': 'The Bigger, Newer Movie',
    'genre': 'Comedy'
})
movies_collection.document('2017_Yet Another Movie').set({
    'year': 2017,
    'title': 'Yet Another Movie',
    'genre': 'Drama'
})
movies_collection.document('2017_One More Movie').set({
    'year': 2017,
    'title': 'One More Movie',
    'genre': 'Thriller'
})
movies_collection.document('2015_An Old Movie').set({
    'year': 2015,
    'title': 'An Old Movie',
    'genre': 'Classic'
})

# --- TODO FIXED: Get a single document by ID ---
print("--- Single Document Fetch ---")
doc_ref = movies_collection.document('2016_The Big New Movie')
doc = doc_ref.get()
if doc.exists:
    print(f"Full Document: {doc.to_dict()}")

# --- TODO FIXED: Get a single document with specific fields only ---
print("\n--- Field Selection (Title & Genre only) ---")
# Use field_paths to restrict the returned data
doc_subset = doc_ref.get(field_paths=['title', 'genre'])
if doc_subset.exists:
    print(f"Subset Data: {doc_subset.to_dict()}")

# --- TODO FIXED: Implement batch document retrieval ---
print("\n--- Batch Retrieval (get_all) ---")
doc_ids = [
    '2016_The Big New Movie',
    '2017_The Bigger, Newer Movie'
]

# Create a list of document references
doc_refs = [movies_collection.document(doc_id) for doc_id in doc_ids]

# Fetch all in one go
docs = db.get_all(doc_refs)

for d in docs:
    if d.exists:
        print(f"Batch Item Found: {d.id} => {d.to_dict()}")
    else:
        print(f"Batch Item Not Found: {d.id}")

```

### Strategic Key Points

* **`field_paths` Optimization**: Note that when you use `field_paths=['title', 'genre']`, the `year` field is completely omitted from the resulting dictionary. This is highly effective for large documents with many fields.
* **The `db.get_all` Advantage**: Using `db.get_all(doc_refs)` is significantly faster than calling `.get()` in a `for` loop because it reduces the number of round-trip requests to the Firestore servers.
* **The `.exists` Guard**: Always check if a document exists before calling `.to_dict()`. If you try to call `.to_dict()` on a non-existent document, your script will throw an error.

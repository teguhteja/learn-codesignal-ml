---

## Understanding GCP Authentication (Production)

Authentication acts as the digital keys to your infrastructure. In GCP, this is typically handled via **Service Accounts** and **Application Default Credentials (ADC)**.

*   **Service Accounts:** These are accounts belonging to your application/VM rather than an individual person.
*   **ADC:** A strategy used by Google libraries to automatically find credentials in your environment.

> **Security Tip:** Never hardcode sensitive keys directly in your scripts. Use environment variables or metadata services.

## Google Cloud SDK Service Discovery

Explore the modular architecture of Google Cloud SDK by running a script that attempts to import several key Google Cloud services and demonstrates the import patterns. This hands-on task will help you understand how Google Cloud SDK's service-specific packages work, showing you the difference between having the meta-package installed versus having actual service libraries available. The script will attempt to import common services like Storage, Compute, and BigQuery, providing practical insight into the import structure and error handling when packages aren't available.

Important Note: Running scripts can alter the filesystem's state or modify the resources in our GCP simulator. To revert to the initial state, you can use the reset button located in the top right corner. However, keep in mind that resetting will erase any code changes. To preserve your code during a reset, consider copying it to the clipboard.

```python
import importlib
import sys

print("Google Cloud SDK - Import Structure Discovery")
print("=" * 55)

# List of common Google Cloud services to test
services_to_test = [
    ("storage", "Cloud Storage", "google-cloud-storage"),
    ("compute", "Compute Engine", "google-cloud-compute"),
    ("bigquery", "BigQuery", "google-cloud-bigquery"),
    ("firestore", "Firestore", "google-cloud-firestore"),
    ("pubsub", "Pub/Sub", "google-cloud-pubsub")
]

print("🔍 Testing Google Cloud Service Imports:")
print("   (Checking availability of common GCP services)")
print()

available_services = []
unavailable_services = []

# Test each service import
for service_name, display_name, package_name in services_to_test:
    try:
        # Attempt to import the service
        module = importlib.import_module(f"google.cloud.{service_name}")
        available_services.append((service_name, display_name, package_name))
        print(f"   ✅ {display_name:<15} | google.cloud.{service_name} | {package_name}")
    except ImportError as e:
        unavailable_services.append((service_name, display_name, package_name))
        print(f"   ❌ {display_name:<15} | google.cloud.{service_name} | {package_name}")

print(f"\n📊 Import Test Results:")
print(f"   • Available services: {len(available_services)}")
print(f"   • Unavailable services: {len(unavailable_services)}")

if available_services:
    print(f"\n✅ Ready to use:")
    for service_name, display_name, package_name in available_services:
        print(f"   from google.cloud import {service_name}")
else:
    print(f"\n⚠️  No Google Cloud services are currently available for import")

if unavailable_services:
    print(f"\n📦 To install missing services:")
    for service_name, display_name, package_name in unavailable_services:
        print(f"   pip install {package_name}")

print("\n" + "=" * 55)
print("🏗️  Google Cloud SDK Architecture Demonstration:")
print("   • Modular Design: Each service is a separate package")
print("   • Independent Installation: Install only what you need")
print("   • Consistent Import Pattern: google.cloud.<service>")
print()
print("💡 Import Pattern Examples:")
print("   from google.cloud import storage    # Cloud Storage client")
print("   from google.cloud import compute    # Compute Engine client")
print("   from google.cloud import bigquery   # BigQuery client")
print()
print("🔧 Installation Pattern Examples:")
print("   pip install google-cloud-storage   # For Cloud Storage")
print("   pip install google-cloud-compute   # For Compute Engine")
print("   pip install google-cloud-bigquery  # For BigQuery")
print("=" * 55)

# Demonstrate the difference between meta-package and service packages
print("\n🎯 Key Learning Points:")
print("   1. google-cloud (meta-package) ≠ service availability")
print("   2. Each service requires its own google-cloud-<service> package")
print("   3. Import errors help identify missing dependencies")
print("   4. Modular approach allows minimal footprint installations")
print("=" * 55)

```

This hands-on task demonstrates the **modular architecture** of the Google Cloud SDK for Python. Unlike monolithic libraries, Google Cloud uses a "plug-and-play" structure where each service is decoupled into its own independent package.



## Understanding the SDK Structure

The Google Cloud SDK is built around the **`google.cloud` namespace**. When you run the script provided, you are exploring how Python handles these sub-packages.

### 1. The Namespace Pattern
All official Google Cloud libraries share a common prefix. This creates a consistent developer experience:
*   **Import:** `from google.cloud import [service]`
*   **Package Name:** `google-cloud-[service]`

### 2. Meta-packages vs. Service-specific Packages
It is a common misconception that installing `pip install google-cloud` is the only way to get started.
*   **`google-cloud` (Meta-package):** This is a "bundle" that acts as a shortcut to install many common services at once. It is often too large for production environments like Cloud Functions or Docker containers.
*   **Service Packages:** Packages like `google-cloud-storage` are lean and contain only the code necessary for that specific service.

## Analysis of the Script

The script uses `importlib.import_module` to dynamically check your environment. This is a practical way to verify your **dependency management**:

*   **Error Handling:** The `try...except ImportError` block mimics what happens in a real application when a dependency is missing from a `requirements.txt` file.
*   **Discovery:** By iterating through `services_to_test`, the script highlights that even if the `google` folder exists in your system, specific sub-folders like `storage` or `bigquery` must be explicitly installed to be functional.

---

## Benefits of Modular Architecture

| Feature | Benefit |
| :--- | :--- |
| **Reduced Footprint** | You only include the code you use, which keeps deployment artifacts (like ZIP files for Lambda or Cloud Functions) small. |
| **Version Independence** | You can update the `storage` library to the latest version without being forced to update `bigquery`, avoiding "dependency hell." |
| **Faster Cold Starts** | In serverless environments, importing fewer modules leads to faster execution starts. |

### Summary of Best Practices
*   **Development:** Use specific packages (`pip install google-cloud-firestore`) to keep your environment clean.
*   **Production:** Always list specific service packages in your `requirements.txt` rather than the broad `google-cloud` meta-package.
*   **Verification:** Use the import pattern `google.cloud.[service]` to ensure your code is using the official, high-level client libraries.

Here is the complete, integrated script. This version combines the **OpenTelemetry configuration**, the **automatic gRPC instrumentation**, and the **custom span logic** into a single production-ready structure.

```python
import os
from opentelemetry import trace
from opentelemetry.sdk.trace import TracerProvider
from opentelemetry.sdk.trace.export import SimpleSpanProcessor, ConsoleSpanExporter
from opentelemetry.sdk.resources import Resource
from opentelemetry.instrumentation.grpc import GrpcInstrumentorClient
from google.cloud import firestore

# --- 1. OPENTELEMETRY SETUP ---
# Define the resource (metadata about your service)
resource = Resource.create({"service.name": "orders-worker"})

# Initialize the Tracer Provider
provider = TracerProvider(resource=resource)
trace.set_tracer_provider(provider)

# Get a tracer instance for this specific module
tracer = trace.get_tracer("OrdersApp")

# Configure the Exporter (Prints traces as JSON to the terminal)
# In production, you would swap this for CloudTraceSpanExporter
console_exporter = ConsoleSpanExporter()

# Add the processor to the provider
# SimpleSpanProcessor is great for debugging; BatchSpanProcessor is for production
provider.add_span_processor(SimpleSpanProcessor(console_exporter))


# --- 2. AUTOMATIC INSTRUMENTATION ---
# MUST be called before initializing the Firestore/GCP clients
GrpcInstrumentorClient().instrument()


# --- 3. GCP CLIENT INITIALIZATION ---
# This client will now automatically generate spans for every network call
db = firestore.Client()


# --- 4. TRACED LOGIC ---
def process_order():
    """
    Simulates fetching a document reference and writing data,
    wrapped in custom spans for granular visibility.
    """
    # High-level parent span for the entire operation
    with tracer.start_as_current_span("process-order-task"):
        
        # Custom span for the reference creation (local logic)
        with tracer.start_as_current_span("get-document-reference"):
            doc_ref = db.collection("orders").document("o-500")
            order_data = {
                "order_id": "o-500", 
                "amount": 199.99, 
                "status": "pending"
            }

        # Custom span for the actual database write (network logic)
        # Note: Auto-instrumentation will create a child span inside this one
        with tracer.start_as_current_span("write-document"):
            doc_ref.set(order_data)

def check_results():
    """Simple query to verify the write, also traced."""
    with tracer.start_as_current_span("verify-order-exists"):
        doc = db.collection("orders").document("o-500").get()
        if doc.exists:
            print(f"Verified: {doc.to_dict()}")

# --- 5. EXECUTION ---
if __name__ == "__main__":
    try:
        process_order()
        check_results()
    except Exception as e:
        print(f"An error occurred: {e}")
    finally:
        # CRITICAL: Ensures all spans are flushed to the console before exiting
        provider.shutdown()
        print("Tracing session complete. Done.")
```

### Why this code works:
1.  **Hierarchy**: By using `with tracer.start_as_current_span("process-order-task")` at the top level, all subsequent spans (both manual and automatic gRPC ones) will be nested under this parent, creating a clear tree structure in your logs.
2.  **Instrumentation Timing**: `GrpcInstrumentorClient().instrument()` is called early. This ensures that when the `firestore.Client()` is created, its internal communication channels are already "tracing-aware."
3.  **Cleanup**: The `provider.shutdown()` in the `finally` block is the safety net that prevents your script from finishing before the last JSON trace has a chance to print to your terminal.

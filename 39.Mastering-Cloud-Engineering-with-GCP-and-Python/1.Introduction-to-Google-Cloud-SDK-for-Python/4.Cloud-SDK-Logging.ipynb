## Introduction to Logging in Cloud SDKs

Welcome to a new adventure! As we continue our journey working with cloud services, this lesson focuses on the essential concept of **logging** within cloud SDKs. 

Logging is a powerful tool that provides visibility into how your application interacts with cloud services. By capturing detailed information about requests, responses, and potential issues, logging enables you to monitor, debug, and optimize your cloud-based operations.

---

## Importance of Logging

Cloud applications can be complex puzzles. Logs act like breadcrumbs that help you navigate these puzzles by providing insights into:
*   **Service Interactions**: Seeing exactly what data is being sent to and from GCP.
*   **Issue Identification**: Pinpointing where a request failed or why it timed out.
*   **Operational Health**: Tracking the overall performance and status of your cloud resources.



---

## Logging in Practice

Most cloud SDKs provide built-in support for logging, allowing you to control the verbosity of the information captured. You can configure logging to capture all activities or focus on specific modules.

Here is an example of how to enable detailed logging for Google Cloud services in Python:

```python
import logging
from google.cloud import firestore

# 1. Set up basic logging configuration
logging.basicConfig(level=logging.INFO)

# 2. Enable detailed DEBUG logging for specific GCP components
# google.cloud: General service logic
# google.auth: Authentication and token exchange
# urllib3.connectionpool: Low-level HTTP requests and responses
logging.getLogger('google.cloud').setLevel(logging.DEBUG)
logging.getLogger('google.auth').setLevel(logging.DEBUG)
logging.getLogger('urllib3.connectionpool').setLevel(logging.DEBUG)

# 3. Interact with the service
client = firestore.Client()
collections = list(client.collections())
print("Collections:", collections)
```

In this example, the `urllib3.connectionpool` logger is particularly useful because it captures the actual HTTP request/response details for API calls, which is invaluable for troubleshooting connectivity.

---

## Recap: Common Logging Levels

Logging systems use severity levels to determine the importance of an event. Most systems ignore messages below a certain threshold (usually `WARNING`) by default.

| Level | Description |
| :--- | :--- |
| **DEBUG** | Detailed information, typically used only for diagnosing problems. |
| **INFO** | Confirmation that things are working as expected. |
| **WARNING** | An indication that something unexpected happened (e.g., "disk space low"). |
| **ERROR** | A serious problem that prevented a specific function from executing. |
| **CRITICAL** | A very serious error that may prevent the application from running. |

**Pro Tip:** It is standard practice to use `DEBUG` or `INFO` during development and switch to `WARNING` or `ERROR` in production to reduce noise and storage costs.

---

## Logging Best Practices and Considerations

Logging requires a balance between visibility and security:
*   **Avoid Sensitive Data**: Never log passwords, access tokens, or Personally Identifiable Information (PII).
*   **Restrict Access**: Ensure that only authorized personnel can view production logs.
*   **Be Descriptive but Concise**: Logs should contain enough context to be useful without creating massive "walls of text" that are hard to search.

## Enable and Observe Firestore Debug Logging with Python

Welcome to your next hands-on task! Here, you have a piece of Python code that initializes a Google Cloud Firestore client, applies GCP debug logging, creates a new collection, and then lists all the available collections. You need to run this script to observe how Python interacts with the local GCP environment and how the debugging logs offer insights into those interactions.

Important Notes:

    The debug logging output can be viewed under the ERRORS tab in your console. This is because the debug level logs are treated as errors in the coding environment. Don't be alarmed, these are not actual errors but helpful debug information for developers.
    Running scripts can modify the resources in our GCP simulator. To revert to the initial state, you can use the reset button located in the top right corner. However, keep in mind that resetting will erase any code changes. To preserve your code during a reset, consider copying it to the clipboard.


```python
import logging
from google.cloud import firestore

# Configure logging to output detailed information for debugging
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s'
)

# Enable detailed logging for Google Cloud services
logging.getLogger('google.cloud').setLevel(logging.DEBUG)
logging.getLogger('google.auth').setLevel(logging.DEBUG)
logging.getLogger('urllib3.connectionpool').setLevel(logging.DEBUG)

# Create a logger for our application
logger = logging.getLogger(__name__)

# Add explicit debug messages to verify logging is working
logger.debug("Starting Firestore operations with debug logging enabled")

# Initialize the Google Cloud Firestore client
client = firestore.Client()

logger.debug("Creating document in Firestore collection")
# Create a new collection as the initial setup is empty
doc_ref = client.collection('my-logging-test-collection').document('test-doc')
doc_ref.set({'message': 'Hello from Firestore!'})

logger.debug("Listing all collections")
# List collections to show logging output
collections = list(client.collections())
print("Collections:")
for collection in collections:
    print(collection.id)
    logger.debug(f"Found collection: {collection.id}")

logger.debug("Firestore operations completed")

```

By running this script, you are engaging with the **observability** layer of cloud development. While your code performs a simple "write and list" operation, the logging configuration reveals the complex "conversation" happening between the Python SDK and the Firestore service.



### What to Look for in the Logs
Since you have set the level to `DEBUG`, the **ERRORS** tab will fill with data. Here is a breakdown of what each logger is telling you:

*   **`google.auth`**: You will see entries related to credential discovery. In a local environment, this often shows the SDK checking for a Service Account key or checking if the `FIRESTORE_EMULATOR_HOST` variable is present.
*   **`urllib3.connectionpool`**: This is the most "chatty" logger. It shows the raw HTTP/gRPC traffic. You’ll see the exact endpoints being called (e.g., `POST http://localhost:8080/...`) and the HTTP status codes returned (like `200 OK`).
*   **`google.cloud`**: This tracks the high-level logic of the Firestore library itself, such as when it starts a stream or prepares a batch write.



### Why logs appear in the "ERRORS" tab
In many cloud-based IDEs and containerized environments (like CodeSignal or Google Cloud Run), the **Standard Error (stderr)** stream is used for all system-level logging, regardless of severity. Because Python's `logging` module defaults to `stderr`, the interface labels these messages as "Errors." 

### Key Benefits of this Setup
*   **Transparency**: You aren't just trusting that `doc_ref.set()` worked; you can actually see the bytes being sent over the wire.
*   **Troubleshooting**: If a connection is refused, the `urllib3` logs will show you exactly which port and IP address the client was trying to reach.
*   **Performance Insight**: By looking at the `asctime` (timestamps) in the logs, you can calculate the latency between your request and the service's response.

Once you have reviewed the output, you’ll have a much better sense of the "invisible" work the Google Cloud SDK does for you every time you initialize a client.

```python
import logging
from google.cloud import firestore

# --- TODO FIXED: Set up logging to capture DEBUG information ---

# 1. Basic configuration for the entire Python logging system
logging.basicConfig(level=logging.DEBUG)

# 2. Specifically target Google Cloud and related network libraries
# 'google.cloud' captures high-level SDK logic
# 'google.auth' captures credential/token exchange
# 'urllib3' captures the raw HTTP request/response details
logging.getLogger('google.cloud').setLevel(logging.DEBUG)
logging.getLogger('google.auth').setLevel(logging.DEBUG)
logging.getLogger('urllib3').setLevel(logging.DEBUG)

# -------------------------------------------------------------

# Initialize the Google Cloud Firestore client
client = firestore.Client()

# Create a new collection with a sample document
doc_ref = client.collection('my-debug-logging-collection').document('sample-doc')
doc_ref.set({'message': 'Hello, debug logging!'})

# Iteratively print all collection names
print("Collections found:")
for collection in client.collections():
    print(f"- {collection.id}")
```



### Why this configuration is important:
*   **The Global Threshold**: `logging.basicConfig(level=logging.DEBUG)` ensures that the application as a whole is ready to listen to debug messages.
*   **Granular Control**: `logging.getLogger('google.cloud').setLevel(logging.DEBUG)` is used because many library authors set their individual loggers to `WARNING` by default to avoid cluttering your console. By explicitly calling `.setLevel()`, you override that default.
*   **Observability**: Once you run this, you will see exactly how the client discovers your emulator or project credentials and the precise moment it sends the data for `sample-doc` to the Firestore service.

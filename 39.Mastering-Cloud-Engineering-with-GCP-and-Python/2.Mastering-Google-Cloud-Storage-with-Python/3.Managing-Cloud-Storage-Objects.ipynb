## Managing Cloud Storage Objects

# Introduction to Cloud Storage Objects

This lesson expands your cloud storage knowledge to include managing **objects** (files) within buckets. We'll explore uploading, downloading, and deleting objects using the client library, focusing on real-world applications such as managing user uploads or maintaining an image archive.

---

## Understanding Blobs

In Cloud Storage, files are represented as **blobs** (Binary Large Objects). A blob is a reference to a file within a bucket. Before performing operations on a file, you must create a blob object to act as a local handle for that cloud resource.

```python
bucket = client.bucket('cosmo-user-uploads')
blob = bucket.blob('cosmo-profile-2023.jpg')  # Creates a blob reference

```

> **Note:** The `blob()` method only creates a local reference. It does **not** check if the file exists in the cloud or start an upload yet.

---

## Uploading Files to Buckets

To upload a file, you need the bucket name, the desired object name, and the local file path. The `upload_from_filename()` method handles the heavy lifting of transferring the bytes from your system to the cloud.

```python
from google.cloud import storage
from google.auth.credentials import AnonymousCredentials

client = storage.Client(
    project=os.environ.get('PROJECT_ID'),
    credentials=AnonymousCredentials(),
    client_options={'api_endpoint': os.environ.get('STORAGE_HOST')}
)

bucket = client.bucket('cosmo-user-uploads')
blob = bucket.blob('cosmo-profile-2023.jpg')

# This transfers the actual file data
blob.upload_from_filename('path/to/local/file.jpg') 

```

---

## Listing Objects in a Bucket

To manage data effectively, you often need to see what is already stored. The `list_blobs()` method returns an iterator of all blob objects within a specific bucket.

```python
# Returns an iterator of all blobs in the bucket
blobs = client.list_blobs('cosmo-user-uploads')  

for blob in blobs:
    print(blob.name)  # Access the object's name in the bucket

```

---

## Downloading Files from Buckets

Downloading follows the same logic as uploading: create a blob reference to the cloud file, then specify where to save it locally using `download_to_filename()`.

```python
bucket = client.bucket('cosmo-user-uploads')
blob = bucket.blob('cosmo-profile-2023.jpg')  # Reference to the cloud file

# Downloads the cloud content to your local path
blob.download_to_filename('local/path/cosmo-profile.jpg')  

```

---

## Retrieving Object Metadata

Objects are more than just raw data; they contain **metadata** like file type, size, and creation date. Use the `reload()` method to ensure your local blob object has the latest info from the server.

```python
bucket = client.bucket('cosmo-images-archive-2023')
blob = bucket.blob('cosmo-profile-2023.jpg')

blob.reload()  # Refreshes metadata from the server

print("Content type:", blob.content_type)  # e.g., 'image/jpeg'
print("Size (bytes):", blob.size)          # File size in bytes
print("Custom metadata:", blob.metadata)   # Custom key-value pairs

```

---

## Deleting Files from Buckets

The `delete()` method permanently removes objects. You can delete single files or iterate through a list to clear a bucket.

```python
# Deleting a single object
bucket = client.bucket('cosmo-user-uploads')
blob = bucket.blob('temporary-file.jpg')
blob.delete()  # Permanently removes the object from the cloud

# Deleting all objects in a bucket (Cleanup)
blobs = client.list_blobs('cosmo-user-uploads')
for blob in blobs:
    blob.delete() 

```

---

## Handling Exceptions Gracefully

Cloud operations can fail if a bucket name is mistyped or a file is missing. Using specific exceptions like `NotFound` ensures your application doesn't crash unexpectedly.

```python
from google.cloud import storage
from google.api_core.exceptions import NotFound, GoogleAPIError

client = storage.Client()

try:
    bucket = client.get_bucket('cosmo-user-uploads')
    bucket.delete()  # Note: Bucket must be empty to be deleted
except NotFound:
    print("Error: The specified bucket does not exist.")
except GoogleAPIError as e:
    print(f"An unexpected API error occurred: {e.message}")

```

### Key Technical Concepts

* **Symmetry in SDK**: You'll notice a distinct symmetry in the Google Cloud Storage SDK. You use `upload_from_filename()` to push data and `download_to_filename()` to pull it. Both require a `blob` object as the bridge.
* **Path Management**: Using `os.path.join` or explicitly defining the full path (including the filename) is essential. If you only provide the directory path, the SDK will throw an error because it needs a specific file handle to write the data.
* **Overwriting**: By default, `download_to_filename()` will overwrite a local file if one already exists at that path with the same name. If you want to prevent this in a real-world app, you would typically check `os.path.exists()` before calling the download method.

### Key Technical Insights

* **The Iterator Pattern**: `list_blobs()` returns an iterator rather than a static list. This is a design choice for **scalability**. If your bucket grows to hold millions of images, your script won't crash by trying to load every filename into memory at once; it will fetch them in "pages" as you loop through them.
* **Blob vs. Object**: In GCS terminology, "Object" is the general cloud resource, while "Blob" is the specific name used in the Python SDK (standing for Binary Large Object).
* **Metadata Availability**: When you list blobs, the SDK automatically retrieves basic metadata for each one (like `.name` and `.size`). This allows you to perform audits or generate reports without needing to call `reload()` on every single file individually.

To master cloud stewardship, you must respect the **Bucket Deletion Rule**: Google Cloud Storage will not allow you to delete a bucket unless it is completely empty. This prevents the accidental loss of potentially thousands of files with a single command.

Here is the logic to perform a "clean sweep" by iterating through the objects before removing the container:

```python
import os
from google.cloud import storage
from google.auth.credentials import AnonymousCredentials
from google.api_core.exceptions import NotFound

# Initialize the Google Cloud Storage client
client = storage.Client(
    project=os.environ.get('PROJECT_ID'),
    credentials=AnonymousCredentials(),
    client_options={'api_endpoint': os.environ.get('STORAGE_HOST')}
)

# Recreating the initial environment
bucket_name = 'codesignal-learn-course-logos'
try:
    bucket = client.get_bucket(bucket_name)
except NotFound:
    bucket = client.create_bucket(bucket_name)

image_files = [
    'prompt-engineering-course-logo.jpg',
    'machine-learning-course-logo.jpg',
    'data-science-python-course-logo.jpg'
]

for image_file in image_files:
    blob = bucket.blob(image_file)
    blob.upload_from_filename(f"/usercode/FILESYSTEM/assets/{image_file}")

# --- TODO FIXED: Clean Sweep Logic ---

print(f"Starting cleanup for bucket: {bucket_name}")

# 1. Delete all objects (blobs) from the bucket
blobs = client.list_blobs(bucket_name)

for blob in blobs:
    print(f"Deleting object: {blob.name}")
    blob.delete()

# 2. Delete the bucket itself now that it is empty
print(f"Deleting empty bucket: {bucket_name}")
bucket.delete()

print("Cleanup complete. Resources decommissioned.")

# -------------------------------------

```

---

### Why this Multi-Step Process is Important

* **Data Protection**: The error thrown when trying to delete a non-empty bucket is a safety feature. It forces you to explicitly acknowledge that you are destroying the data inside.
* **The Iterator Pattern**: When you call `list_blobs()`, the SDK fetches the metadata. Calling `.delete()` on each `blob` object tells the API to remove that specific entry from the cloud index.
* **Finality**: Once `bucket.delete()` is called, the globally unique name `codesignal-learn-course-logos` is released. Anyone else in the world (or you, in a different project) can now claim that name.

### Pro-Tip: Bulk Deletion

In a production environment with millions of files, deleting them one by one in a loop can be slow. For massive cleanups, developers often use **Bucket Lifecycle Management** rules to expire files automatically or use the `gsutil rm -r gs://bucket-name` command-line tool, which handles the concurrent deletion logic much faster than a sequential Python loop.

This challenge represents the full "CRUD" (Create, Read, Update, Delete) lifecycle of cloud infrastructure management. Mastering this flow is essential for ensuring that your cloud environment remains cost-effective and organized.

Here is the complete solution for the Full Cycle Management Challenge:

```python
import os
from google.cloud import storage
from google.auth.credentials import AnonymousCredentials
from google.api_core import exceptions

# Initialize the Google Cloud Storage client
client = storage.Client(
    project=os.environ.get('PROJECT_ID'),
    credentials=AnonymousCredentials(),
    client_options={'api_endpoint': os.environ.get('STORAGE_HOST')}
)

bucket_name = 'full-cycle-management-challenge'

# 1. Create the bucket using the bucket_name variable
try:
    bucket = client.create_bucket(bucket_name)
    print(f"Phase 1: Bucket '{bucket_name}' created.")
except exceptions.Conflict:
    bucket = client.get_bucket(bucket_name)
    print(f"Phase 1: Bucket '{bucket_name}' already exists.")

# Names of the images to be uploaded
images = [
    'prompt-engineering-course-logo.jpg',
    'machine-learning-course-logo.jpg',
    'data-science-python-course-logo.jpg'
]

# 2. Upload the images to the bucket from '/usercode/FILESYSTEM/assets/'
print("\nPhase 2: Uploading assets...")
for image_name in images:
    blob = bucket.blob(image_name)
    local_path = f'/usercode/FILESYSTEM/assets/{image_name}'
    blob.upload_from_filename(local_path)
    print(f" - Uploaded: {image_name}")

# 3. List the contents of the bucket
print("\nPhase 3: Verifying bucket contents:")
blobs = list(client.list_blobs(bucket_name))
for blob in blobs:
    print(f" - Found Object: {blob.name}")

# 4. Delete all objects from the bucket
print("\nPhase 4: Cleaning up objects...")
for blob in blobs:
    blob.delete()
    print(f" - Deleted: {blob.name}")

# 5. Delete the bucket itself
print("\nPhase 5: Decommissioning bucket...")
bucket.delete()
print(f"Success: Bucket '{bucket_name}' has been removed.")

```

### The Life Cycle in Review

1. **Provisioning**: You established a globally unique namespace and allocated cloud resources.
2. **Ingestion**: You moved local binary data into the cloud, transforming files into **Blobs**.
3. **Auditing**: You used `list_blobs` to verify the state of your resources, a critical step in automated pipelines.
4. **Teardown**: You followed the "inside-out" deletion rule, ensuring the bucket was empty before releasing the container.

By completing this cycle, you've demonstrated the ability to programmatically control the birth, life, and retirement of cloud storage resources.
